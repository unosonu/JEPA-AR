```python
"""
JEPA-Guided AR Transformer  —  v4
===================================
Core architectural change from v3:
  JEPA guidance is injected into the REPRESENTATION, not the logits.

  v1/v2/v3:  backbone(ctx) → ar_logits + guidance_head(jepa_emb)
             Problem: guidance head corrects logits of a backbone it never
             trained with. Weak residual signal on a moving target.

  v4:        backbone(ctx) → h
             jepa_emb = jepa_proj(h.detach())          # still detached
             h_guided = h + GuidanceXAttn(h, jepa_emb) # inject BEFORE ar_head
             ar_logits = ar_head(h_guided)

  The GuidanceXAttn layer is a lightweight cross-attention:
    queries = h            (backbone repr of current context)
    keys/values = jepa_emb (projected future embedding, tiled to seq_len)

  This means:
    - Every token position can attend to the JEPA future summary
    - The correction happens in representation space, not logit space
    - Backbone still gets clean AR gradients (h is detached before jepa_proj)
    - GuidanceXAttn trains end-to-end with AR loss — no separate guidance loss needed
    - At inference with use_guidance=False: GuidanceXAttn output is zeroed

  Training losses:
    AR loss   → backbone + ar_head + GuidanceXAttn   (end-to-end)
    JEPA loss → jepa_proj only                        (backbone detached)

Dataset : WikiText-2 (~2MB, auto-downloaded)
Hardware: Free Colab T4 GPU (~6-8 min) | CPU laptop (~25 min)

Install:
    pip install torch datasets tokenizers matplotlib tqdm

Run:
    python jepa_ar_v4.py
"""

import os, math, time, json, random
from copy import deepcopy
from dataclasses import dataclass, field

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
from tqdm import tqdm

# ── reproducibility ───────────────────────────────────────────────────────────
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

# ─────────────────────────────────────────────────────────────────────────────
# 1.  CONFIG
# ─────────────────────────────────────────────────────────────────────────────
@dataclass
class Config:
    # model
    vocab_size       : int   = 8000
    seq_len          : int   = 128
    d_model          : int   = 128
    n_heads          : int   = 4
    n_layers         : int   = 4
    d_ff             : int   = 512
    dropout          : float = 0.1

    # JEPA
    jepa_dim         : int   = 64
    jepa_ema_decay   : float = 0.996
    jepa_lambda      : float = 1.0

    # VICReg
    vicreg_mu        : float = 1.0
    vicreg_nu        : float = 0.04
    vicreg_eps       : float = 1e-4

    # guidance cross-attention
    # Projects jepa_dim → d_model for cross-attn keys/values
    # Starts at zero influence (guidance_init_scale=0) and learns from scratch
    guidance_n_heads : int   = 2         # lightweight, fewer heads than backbone
    guidance_init_scale : float = 0.0    # zero-init output proj → starts neutral

    # training
    batch_size       : int   = 48
    lr               : float = 3e-4
    epochs           : int   = 10
    grad_clip        : float = 1.0
    warmup_steps     : int   = 300

    # eval / logging
    eval_every       : int   = 200
    gen_every        : int   = 400
    gen_prompts      : list  = field(default_factory=lambda: [
        "The history of science",
        "In the early 20th century",
        "The political system of",
        "During the Second World",
    ])
    device : str = "cuda" if torch.cuda.is_available() else "cpu"

CFG = Config()
print(f"Device: {CFG.device}")


# ─────────────────────────────────────────────────────────────────────────────
# 2.  DATA
# ─────────────────────────────────────────────────────────────────────────────

def load_wikitext2():
    from datasets import load_dataset
    ds = load_dataset("wikitext", "wikitext-2-raw-v1")
    return "\n".join(ds["train"]["text"]), "\n".join(ds["validation"]["text"])


def build_tokenizer(text, vocab_size=8000):
    from tokenizers import Tokenizer
    from tokenizers.models import BPE
    from tokenizers.trainers import BpeTrainer
    from tokenizers.pre_tokenizers import Whitespace
    import tempfile
    tok = Tokenizer(BPE(unk_token="[UNK]"))
    tok.pre_tokenizer = Whitespace()
    trainer = BpeTrainer(vocab_size=vocab_size,
                         special_tokens=["[PAD]","[UNK]","[BOS]","[EOS]"])
    with tempfile.NamedTemporaryFile("w", suffix=".txt", delete=False) as f:
        f.write(text); tmp = f.name
    tok.train([tmp], trainer); os.unlink(tmp)
    return tok


def tokenize_and_chunk(text, tokenizer, seq_len):
    ids = tokenizer.encode(text).ids
    ids = ids[:(len(ids) // seq_len) * seq_len]
    return np.array(ids, dtype=np.int32).reshape(-1, seq_len)


class PairDataset(Dataset):
    """Returns (ctx_chunk[i], future_chunk[i+1]) pairs."""
    def __init__(self, chunks):
        self.chunks = torch.tensor(chunks, dtype=torch.long)

    def __len__(self):
        return len(self.chunks) - 1

    def __getitem__(self, i):
        return self.chunks[i], self.chunks[i + 1]


# ─────────────────────────────────────────────────────────────────────────────
# 3.  BACKBONE COMPONENTS
# ─────────────────────────────────────────────────────────────────────────────

class RotaryEmbedding(nn.Module):
    def __init__(self, dim):
        super().__init__()
        inv_freq = 1.0 / (10000 ** (torch.arange(0, dim, 2).float() / dim))
        self.register_buffer("inv_freq", inv_freq)

    def forward(self, seq_len, device):
        t     = torch.arange(seq_len, device=device).float()
        freqs = torch.outer(t, self.inv_freq)
        emb   = torch.cat([freqs, freqs], dim=-1)
        return emb.cos(), emb.sin()


def rotate_half(x):
    x1, x2 = x.chunk(2, dim=-1)
    return torch.cat([-x2, x1], dim=-1)


def apply_rope(q, k, cos, sin):
    return q * cos + rotate_half(q) * sin, k * cos + rotate_half(k) * sin


class MultiHeadSelfAttention(nn.Module):
    def __init__(self, d_model, n_heads, dropout, causal=True):
        super().__init__()
        self.n_heads = n_heads
        self.d_head  = d_model // n_heads
        self.causal  = causal
        self.qkv     = nn.Linear(d_model, 3 * d_model, bias=False)
        self.out     = nn.Linear(d_model, d_model, bias=False)
        self.drop    = nn.Dropout(dropout)
        self.rope    = RotaryEmbedding(self.d_head)

    def forward(self, x):
        B, T, C = x.shape
        cos, sin = self.rope(T, x.device)
        cos = cos.unsqueeze(0).unsqueeze(0)
        sin = sin.unsqueeze(0).unsqueeze(0)
        qkv = self.qkv(x).reshape(B, T, 3, self.n_heads, self.d_head)
        q, k, v = qkv.unbind(2)
        q = q.transpose(1, 2); k = k.transpose(1, 2); v = v.transpose(1, 2)
        q, k = apply_rope(q, k, cos, sin)
        scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(self.d_head)
        if self.causal:
            mask = torch.triu(torch.ones(T, T, device=x.device), diagonal=1).bool()
            scores = scores.masked_fill(mask, float("-inf"))
        attn = self.drop(F.softmax(scores, dim=-1))
        out  = torch.matmul(attn, v).transpose(1, 2).reshape(B, T, C)
        return self.out(out)


class FFN(nn.Module):
    def __init__(self, d_model, d_ff, dropout):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_model, d_ff), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(d_ff, d_model), nn.Dropout(dropout))

    def forward(self, x):
        return self.net(x)


class TransformerBlock(nn.Module):
    def __init__(self, d_model, n_heads, d_ff, dropout):
        super().__init__()
        self.attn = MultiHeadSelfAttention(d_model, n_heads, dropout)
        self.ffn  = FFN(d_model, d_ff, dropout)
        self.ln1  = nn.LayerNorm(d_model)
        self.ln2  = nn.LayerNorm(d_model)

    def forward(self, x):
        x = x + self.attn(self.ln1(x))
        x = x + self.ffn(self.ln2(x))
        return x


# ─────────────────────────────────────────────────────────────────────────────
# 4.  GUIDANCE CROSS-ATTENTION  (the key new component)
# ─────────────────────────────────────────────────────────────────────────────

class GuidanceCrossAttention(nn.Module):
    """
    Lightweight cross-attention layer that injects JEPA future context
    into the backbone representation BEFORE the AR head.

    queries : backbone repr h         (B, T, d_model)
    keys    : jepa future summary     (B, 1, d_model)  ← tiled across positions
    values  : jepa future summary     (B, 1, d_model)

    Each token position queries the single JEPA summary vector.
    The output is a residual correction to h in representation space.

    Zero-initialised output projection → starts as identity (no effect),
    gradually learns to use JEPA context as training progresses.

    At inference with use_guidance=False: returns zeros (no effect on h).
    """

    def __init__(self, d_model, jepa_dim, n_heads, dropout, init_scale=0.0):
        super().__init__()
        assert d_model % n_heads == 0
        self.n_heads = n_heads
        self.d_head  = d_model // n_heads
        self.scale   = math.sqrt(self.d_head)

        # project JEPA embedding → d_model for keys and values
        self.jepa_to_kv = nn.Linear(jepa_dim, 2 * d_model, bias=False)

        # queries come from backbone repr
        self.q_proj  = nn.Linear(d_model, d_model, bias=False)

        # output projection — zero-init so layer starts neutral
        self.out_proj = nn.Linear(d_model, d_model, bias=False)
        nn.init.constant_(self.out_proj.weight, init_scale)

        self.ln   = nn.LayerNorm(d_model)   # pre-norm on queries
        self.drop = nn.Dropout(dropout)

    def forward(self, h, jepa_summary):
        """
        h            : (B, T, d_model)   backbone repr
        jepa_summary : (B, d_model)      mean-pooled JEPA embedding, projected

        Returns correction: (B, T, d_model)  — added to h as residual
        """
        B, T, C = h.shape

        # queries from backbone repr (pre-normed)
        q = self.q_proj(self.ln(h))                        # (B, T, d_model)
        q = q.reshape(B, T, self.n_heads, self.d_head).transpose(1, 2)

        # keys & values from JEPA summary — shape (B, 1, 2*d_model)
        kv = self.jepa_to_kv(jepa_summary).unsqueeze(1)   # (B, 1, 2*d_model)
        k, v = kv.chunk(2, dim=-1)                         # (B, 1, d_model) each
        k = k.reshape(B, 1, self.n_heads, self.d_head).transpose(1, 2)
        v = v.reshape(B, 1, self.n_heads, self.d_head).transpose(1, 2)

        # cross-attention: each query token attends to the single JEPA key
        # scores: (B, n_heads, T, 1)
        scores = torch.matmul(q, k.transpose(-2, -1)) / self.scale
        attn   = self.drop(F.softmax(scores, dim=-1))

        # output: (B, n_heads, T, d_head) → (B, T, d_model)
        out = torch.matmul(attn, v).transpose(1, 2).reshape(B, T, C)
        return self.out_proj(out)


# ─────────────────────────────────────────────────────────────────────────────
# 5.  VICReg
# ─────────────────────────────────────────────────────────────────────────────

def vicreg_loss(z, cfg):
    if z.dim() == 3:
        B, T, D = z.shape
        z = z.reshape(B * T, D)
    z = z - z.mean(dim=0)
    N, D = z.shape
    std      = torch.sqrt(z.var(dim=0) + cfg.vicreg_eps)
    var_loss = torch.relu(1.0 - std).mean()
    cov      = (z.T @ z) / (N - 1)
    off_diag = cov * (1 - torch.eye(D, device=z.device))
    cov_loss = off_diag.pow(2).sum() / D
    return cfg.vicreg_mu * var_loss + cfg.vicreg_nu * cov_loss


# ─────────────────────────────────────────────────────────────────────────────
# 6.  FULL MODEL
# ─────────────────────────────────────────────────────────────────────────────

class JEPAGuidedTransformer(nn.Module):
    """
    Forward pass (training, future provided):

      ctx → backbone → h                           (B, T, d_model)
      h.detach() → jepa_proj → jepa_pred           (B, T, jepa_dim)
      future → target_encoder → target_proj → jepa_target  (B, T, jepa_dim)

      jepa_summary = mean_pool(jepa_pred).detach() (B, jepa_dim)
                     ↑ detached: guidance xattn gets no JEPA gradients
      h_guided = h + GuidanceCrossAttn(h, jepa_summary)
      ar_logits = ar_head(h_guided)

    Gradient flows:
      AR loss (CE on ar_logits) → backbone + GuidanceCrossAttn + ar_head
      JEPA loss                 → jepa_proj only  (h detached before jepa_proj)
      No gradient crosses between the two paths.

    At inference with use_guidance=False:
      h_guided = h  (GuidanceCrossAttn bypassed entirely)
    """

    def __init__(self, cfg: Config):
        super().__init__()
        self.cfg = cfg

        # ── shared backbone ───────────────────────────────────────────────
        self.embed    = nn.Embedding(cfg.vocab_size, cfg.d_model)
        self.drop     = nn.Dropout(cfg.dropout)
        self.blocks   = nn.ModuleList([
            TransformerBlock(cfg.d_model, cfg.n_heads, cfg.d_ff, cfg.dropout)
            for _ in range(cfg.n_layers)
        ])
        self.ln_final = nn.LayerNorm(cfg.d_model)

        # ── AR head (weight-tied) ─────────────────────────────────────────
        self.ar_head  = nn.Linear(cfg.d_model, cfg.vocab_size, bias=False)
        self.ar_head.weight = self.embed.weight

        # ── JEPA online projection (backbone always detached here) ────────
        self.jepa_proj = nn.Sequential(
            nn.Linear(cfg.d_model, cfg.d_model),
            nn.GELU(),
            nn.Linear(cfg.d_model, cfg.jepa_dim),
        )

        # ── JEPA target encoder (EMA, processes future chunk) ─────────────
        self.target_embed  = deepcopy(self.embed)
        self.target_blocks = nn.ModuleList([deepcopy(b) for b in self.blocks])
        self.target_ln     = deepcopy(self.ln_final)
        self.target_proj   = deepcopy(self.jepa_proj)
        for p in self._target_params():
            p.requires_grad_(False)

        # ── Guidance cross-attention (NEW — sits between backbone & ar_head)
        # Projects jepa_dim → d_model internally, queries backbone repr
        self.guidance_xattn = GuidanceCrossAttention(
            d_model    = cfg.d_model,
            jepa_dim   = cfg.jepa_dim,
            n_heads    = cfg.guidance_n_heads,
            dropout    = cfg.dropout,
            init_scale = cfg.guidance_init_scale,
        )

        self._init_weights()

    def _target_params(self):
        for m in [self.target_embed, self.target_ln,
                  self.target_proj, *self.target_blocks]:
            yield from m.parameters()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, std=0.02)
                if m.bias is not None: nn.init.zeros_(m.bias)
            elif isinstance(m, nn.Embedding):
                nn.init.normal_(m.weight, std=0.02)
        # re-apply zero init on guidance output proj (overrides normal init)
        nn.init.constant_(self.guidance_xattn.out_proj.weight,
                          self.cfg.guidance_init_scale)

    @torch.no_grad()
    def update_ema(self):
        d = self.cfg.jepa_ema_decay
        for online, target in [
            (self.embed,     self.target_embed),
            (self.ln_final,  self.target_ln),
            (self.jepa_proj, self.target_proj),
        ]:
            for po, pt in zip(online.parameters(), target.parameters()):
                pt.data.mul_(d).add_(po.data, alpha=1 - d)
        for bo, bt in zip(self.blocks, self.target_blocks):
            for po, pt in zip(bo.parameters(), bt.parameters()):
                pt.data.mul_(d).add_(po.data, alpha=1 - d)

    def _encode_backbone(self, x):
        h = self.drop(self.embed(x))
        for b in self.blocks: h = b(h)
        return self.ln_final(h)

    @torch.no_grad()
    def _encode_target(self, x):
        h = self.target_embed(x)
        for b in self.target_blocks: h = b(h)
        return self.target_ln(h)

    def forward(self, ctx, future=None, use_guidance=True):
        """
        ctx    : (B, T) context tokens
        future : (B, T) future tokens for JEPA target (None at inference)
        use_guidance: if False, bypass GuidanceCrossAttn entirely

        Returns:
            ar_logits   : (B, T, V)
            jepa_pred   : (B, T, jepa_dim)
            jepa_target : (B, T, jepa_dim) or None
        """
        # ── backbone (clean AR gradients flow here) ───────────────────────
        h = self._encode_backbone(ctx)                   # (B, T, d_model)

        # ── JEPA online path (backbone detached) ──────────────────────────
        jepa_pred = self.jepa_proj(h.detach())           # (B, T, jepa_dim)

        # ── JEPA target (EMA future encoder) ──────────────────────────────
        jepa_target = None
        if future is not None:
            with torch.no_grad():
                h_fut       = self._encode_target(future)
                jepa_target = self.target_proj(h_fut)    # (B, T, jepa_dim)

        # ── Guidance cross-attention (representation-level injection) ──────
        if use_guidance:
            # summarise JEPA future prediction as a single vector per batch item
            # detach so guidance xattn grads don't flow back to jepa_proj
            jepa_summary = jepa_pred.mean(dim=1).detach()  # (B, jepa_dim)
            correction   = self.guidance_xattn(h, jepa_summary)
            h_guided     = h + correction                   # residual add
        else:
            h_guided = h   # clean AR, no guidance

        # ── AR head ───────────────────────────────────────────────────────
        ar_logits = self.ar_head(h_guided)               # (B, T, V)

        return ar_logits, jepa_pred, jepa_target

    @torch.no_grad()
    def generate(self, prompt_ids, max_new=60, temperature=0.8,
                 top_k=40, use_guidance=True):
        self.eval()
        device = next(self.parameters()).device
        ids    = prompt_ids.to(device)
        for _ in range(max_new):
            ctx = ids[:, -self.cfg.seq_len:]
            ar_logits, _, _ = self.forward(ctx, future=None,
                                           use_guidance=use_guidance)
            logits = ar_logits[:, -1, :] / temperature
            if top_k > 0:
                topk_v, _ = torch.topk(logits, top_k)
                logits[logits < topk_v[:, -1:]] = float("-inf")
            ids = torch.cat([ids,
                             torch.multinomial(F.softmax(logits, -1), 1)], dim=1)
        self.train()
        return ids[0].tolist()


# ─────────────────────────────────────────────────────────────────────────────
# 7.  LOSS
# ─────────────────────────────────────────────────────────────────────────────

def compute_loss(ar_logits, jepa_pred, jepa_target, ctx_tokens, cfg):
    """
    AR loss  : CE on ar_logits (guided repr) → backbone + xattn + ar_head
    JEPA loss: alignment + VICReg            → jepa_proj only
    """
    B, T, V = ar_logits.shape

    # AR loss
    ar_loss = F.cross_entropy(
        ar_logits[:, :-1, :].reshape(-1, V),
        ctx_tokens[:, 1:].reshape(-1),
        ignore_index=0,
    )

    # JEPA loss
    jepa_loss = torch.tensor(0.0, device=ar_logits.device)
    align_val = vc_val = 0.0
    if jepa_target is not None:
        pred_vec   = F.normalize(jepa_pred[:, -1, :],     dim=-1)
        target_vec = F.normalize(jepa_target.mean(dim=1), dim=-1)
        align_loss = F.smooth_l1_loss(pred_vec, target_vec)
        vc_loss    = vicreg_loss(jepa_pred, cfg)
        jepa_loss  = align_loss + vc_loss
        align_val  = align_loss.item()
        vc_val     = vc_loss.item()

    total = ar_loss + cfg.jepa_lambda * jepa_loss
    return total, ar_loss.item(), align_val, vc_val


# ─────────────────────────────────────────────────────────────────────────────
# 8.  TRAINING UTILITIES
# ─────────────────────────────────────────────────────────────────────────────

def get_lr(step, cfg):
    if step < cfg.warmup_steps:
        return cfg.lr * step / max(1, cfg.warmup_steps)
    return cfg.lr


@torch.no_grad()
def evaluate(model, loader, cfg, max_batches=50):
    model.eval()
    tot = dict(ar=0, align=0, vc=0, n=0)
    for i, (ctx, future) in enumerate(loader):
        if i >= max_batches: break
        ctx, future = ctx.to(cfg.device), future.to(cfg.device)
        ar_logits, jp, jt = model(ctx, future, use_guidance=True)
        _, ar_l, al_l, vc_l = compute_loss(ar_logits, jp, jt, ctx, cfg)
        tot["ar"] += ar_l; tot["align"] += al_l; tot["vc"] += vc_l; tot["n"] += 1
    model.train()
    n = tot["n"]
    return tot["ar"]/n, tot["align"]/n, tot["vc"]/n


@torch.no_grad()
def evaluate_both(model, loader, cfg, max_batches=50):
    """Evaluate PPL with and without guidance — shows guidance benefit directly."""
    model.eval()
    ar_guided = ar_base = n = 0
    for i, (ctx, future) in enumerate(loader):
        if i >= max_batches: break
        ctx, future = ctx.to(cfg.device), future.to(cfg.device)
        B, T = ctx.shape
        V = cfg.vocab_size

        logits_g, _, _ = model(ctx, future, use_guidance=True)
        logits_b, _, _ = model(ctx, future, use_guidance=False)

        ar_guided += F.cross_entropy(
            logits_g[:, :-1].reshape(-1, V), ctx[:, 1:].reshape(-1),
            ignore_index=0).item()
        ar_base += F.cross_entropy(
            logits_b[:, :-1].reshape(-1, V), ctx[:, 1:].reshape(-1),
            ignore_index=0).item()
        n += 1
    model.train()
    return math.exp(ar_guided / n), math.exp(ar_base / n)


def decode(ids, tokenizer):
    return tokenizer.decode(ids)


def print_generation(model, tokenizer, cfg, step):
    print(f"\n  ── Generation at step {step} ──")
    for prompt in cfg.gen_prompts[:2]:
        pid = torch.tensor([tokenizer.encode(prompt).ids], dtype=torch.long)
        with_g    = decode(model.generate(pid, max_new=50, use_guidance=True),  tokenizer)
        without_g = decode(model.generate(pid, max_new=50, use_guidance=False), tokenizer)
        print(f"\n  Prompt : \"{prompt}\"")
        print(f"  [WITH]   {with_g}")
        print(f"  [WITHOUT]{without_g}")


# ─────────────────────────────────────────────────────────────────────────────
# 9.  MAIN TRAINING LOOP
# ─────────────────────────────────────────────────────────────────────────────

def train(cfg):
    print("\n── Loading WikiText-2 ──────────────────────────────────────")
    train_text, val_text = load_wikitext2()
    print(f"  Train chars: {len(train_text):,}   Val chars: {len(val_text):,}")

    print("\n── Building tokenizer ─────────────────────────────────────")
    tokenizer = build_tokenizer(train_text, cfg.vocab_size)
    cfg.vocab_size = tokenizer.get_vocab_size()
    print(f"  Vocab size: {cfg.vocab_size}")

    print("\n── Tokenising & chunking ──────────────────────────────────")
    train_chunks = tokenize_and_chunk(train_text, tokenizer, cfg.seq_len)
    val_chunks   = tokenize_and_chunk(val_text,   tokenizer, cfg.seq_len)
    print(f"  Train pairs: {len(train_chunks)-1:,}   Val pairs: {len(val_chunks)-1:,}")

    train_loader = DataLoader(PairDataset(train_chunks),
                              batch_size=cfg.batch_size, shuffle=True,  num_workers=0)
    val_loader   = DataLoader(PairDataset(val_chunks),
                              batch_size=cfg.batch_size, shuffle=False, num_workers=0)

    print("\n── Building model ─────────────────────────────────────────")
    model = JEPAGuidedTransformer(cfg).to(cfg.device)
    n_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"  Trainable parameters: {n_params:,}")
    print(f"\n  Architecture:")
    print(f"    backbone    → h")
    print(f"    h.detach()  → jepa_proj  → jepa_pred  (JEPA loss only)")
    print(f"    jepa_pred.mean().detach() → GuidanceCrossAttn(h, ·) → correction")
    print(f"    h + correction → ar_head → ar_logits  (AR loss flows here)")

    optimizer = torch.optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=cfg.lr, weight_decay=0.01)

    history = dict(step=[], train_ar=[], train_align=[], train_vc=[],
                   val_ar=[], val_align=[], val_vc=[],
                   val_ppl_guided=[], val_ppl_base=[],
                   train_ppl=[], val_ppl=[])

    step = 0; best_val_ppl = float("inf")
    ep_ar = ep_align = ep_vc = ep_n = 0

    print("\n── Training ────────────────────────────────────────────────")

    for epoch in range(1, cfg.epochs + 1):
        ep_ar = ep_align = ep_vc = ep_n = 0
        pbar  = tqdm(train_loader, desc=f"Epoch {epoch}/{cfg.epochs}")

        for ctx, future in pbar:
            ctx, future = ctx.to(cfg.device), future.to(cfg.device)
            for pg in optimizer.param_groups: pg["lr"] = get_lr(step, cfg)

            optimizer.zero_grad()
            ar_logits, jp, jt = model(ctx, future, use_guidance=True)
            loss, ar_l, al_l, vc_l = compute_loss(ar_logits, jp, jt, ctx, cfg)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.grad_clip)
            optimizer.step()
            model.update_ema()

            ep_ar += ar_l; ep_align += al_l; ep_vc += vc_l
            ep_n  += 1; step += 1

            pbar.set_postfix({
                "ar":    f"{ar_l:.3f}",
                "align": f"{al_l:.4f}",
                "vc":    f"{vc_l:.4f}",
                "ppl":   f"{math.exp(ar_l):.1f}",
            })

            if step % cfg.eval_every == 0:
                val_ar, val_al, val_vc = evaluate(model, val_loader, cfg)
                ppl_guided, ppl_base   = evaluate_both(model, val_loader, cfg)
                train_ppl = math.exp(ep_ar / ep_n)

                history["step"].append(step)
                history["train_ar"].append(ep_ar / ep_n)
                history["train_align"].append(ep_align / ep_n)
                history["train_vc"].append(ep_vc / ep_n)
                history["val_ar"].append(val_ar)
                history["val_align"].append(val_al)
                history["val_vc"].append(val_vc)
                history["val_ppl_guided"].append(ppl_guided)
                history["val_ppl_base"].append(ppl_base)
                history["train_ppl"].append(train_ppl)
                history["val_ppl"].append(ppl_guided)

                if ppl_guided < best_val_ppl:
                    best_val_ppl = ppl_guided
                    torch.save(model.state_dict(), "best_model_v4.pt")

                delta = ppl_base - ppl_guided
                print(f"\n  Step {step:5d} | "
                      f"Train PPL {train_ppl:.1f} | "
                      f"Val PPL guided {ppl_guided:.1f} | "
                      f"base {ppl_base:.1f} | "
                      f"Δ {delta:+.2f} | "
                      f"align {val_al:.4f} | vc {val_vc:.4f}")

            if step % cfg.gen_every == 0:
                print_generation(model, tokenizer, cfg, step)

    return model, tokenizer, history


# ─────────────────────────────────────────────────────────────────────────────
# 10. ABLATION  —  AR-only vs JEPA-guided v4 (6 epochs — enough to stabilise)
# ─────────────────────────────────────────────────────────────────────────────

def run_ablation(train_loader, val_loader, tokenizer, cfg, epochs=6):
    print(f"\n── Ablation: AR-only vs JEPA-guided v4 ({epochs} epochs) ───")

    def train_variant(use_jepa, label):
        cfg2 = deepcopy(cfg)
        if not use_jepa:
            cfg2.jepa_lambda = 0.0

        m   = JEPAGuidedTransformer(cfg2).to(cfg.device)
        opt = torch.optim.AdamW(
            [p for p in m.parameters() if p.requires_grad],
            lr=cfg2.lr, weight_decay=0.01)

        ppls_guided = []
        ppls_base   = []

        for ep in range(epochs):
            for ctx, future in tqdm(train_loader,
                                    desc=f"  {label} ep{ep+1}", leave=False):
                ctx, future = ctx.to(cfg.device), future.to(cfg.device)
                opt.zero_grad()
                ar_logits, jp, jt = m(ctx, future, use_guidance=use_jepa)
                loss, _, _, _ = compute_loss(ar_logits, jp, jt, ctx, cfg2)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(m.parameters(), cfg2.grad_clip)
                opt.step(); m.update_ema()

            if use_jepa:
                pg, pb = evaluate_both(m, val_loader, cfg2)
            else:
                pb, _ = evaluate_both(m, val_loader, cfg2)
                pg = pb
            ppls_guided.append(pg)
            ppls_base.append(pb)
            print(f"    {label}  epoch {ep+1}  "
                  f"PPL guided={pg:.2f}  base={pb:.2f}  Δ={pb-pg:+.2f}")
        return ppls_guided, ppls_base

    ar_ppls, _          = train_variant(False, "AR-only  ")
    jepa_guided, jepa_base = train_variant(True,  "JEPA+AR  ")

    # final numbers
    delta_vs_base = jepa_base[-1] - jepa_guided[-1]   # guidance benefit within JEPA model
    delta_vs_ar   = ar_ppls[-1]   - jepa_guided[-1]   # JEPA guided vs pure AR

    print(f"\n  ── Final results (epoch {epochs}) ──")
    print(f"  AR-only PPL              : {ar_ppls[-1]:.2f}")
    print(f"  JEPA model (no guidance) : {jepa_base[-1]:.2f}")
    print(f"  JEPA model (guided)      : {jepa_guided[-1]:.2f}")
    print(f"  Guidance benefit (Δ)     : {delta_vs_base:+.2f}  "
          f"({'✓ guidance helps' if delta_vs_base > 0 else '✗ guidance hurts'})")
    print(f"  vs AR-only (Δ)           : {delta_vs_ar:+.2f}  "
          f"({'✓ JEPA beats AR' if delta_vs_ar > 0 else '✗ AR still better'})")

    # ── plot ──────────────────────────────────────────────────────────────
    ep_range = list(range(1, epochs + 1))
    fig, axes = plt.subplots(1, 3, figsize=(16, 4))
    fig.suptitle("v4 Ablation: AR-only vs JEPA-Guided", fontsize=13)

    axes[0].plot(ep_range, ar_ppls,     "o-", label="AR only",            color="#E05C5C", lw=2)
    axes[0].plot(ep_range, jepa_guided, "s-", label="JEPA guided",         color="#4A90D9", lw=2)
    axes[0].plot(ep_range, jepa_base,   "^--",label="JEPA no guidance",    color="#85C1E9", lw=1.5)
    axes[0].set_title("Val PPL (lower = better)")
    axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Val PPL")
    axes[0].legend(fontsize=8); axes[0].grid(True, alpha=0.3)

    # guidance benefit per epoch (base - guided within JEPA model)
    deltas_guidance = [b - g for b, g in zip(jepa_base, jepa_guided)]
    colors = ["#4A90D9" if d > 0 else "#E05C5C" for d in deltas_guidance]
    axes[1].bar(ep_range, deltas_guidance, color=colors)
    axes[1].axhline(0, color="black", lw=0.8)
    axes[1].set_title("Guidance benefit per epoch\n(JEPA_base − JEPA_guided, +ve = helps)")
    axes[1].set_xlabel("Epoch"); axes[1].grid(True, alpha=0.3, axis="y")

    # JEPA vs AR delta
    deltas_vs_ar = [ar - jg for ar, jg in zip(ar_ppls, jepa_guided)]
    colors2 = ["#4A90D9" if d > 0 else "#E05C5C" for d in deltas_vs_ar]
    axes[2].bar(ep_range, deltas_vs_ar, color=colors2)
    axes[2].axhline(0, color="black", lw=0.8)
    axes[2].set_title("JEPA-guided vs AR-only\n(AR − JEPA_guided, +ve = JEPA wins)")
    axes[2].set_xlabel("Epoch"); axes[2].grid(True, alpha=0.3, axis="y")

    plt.tight_layout()
    plt.savefig("ablation_v4.png", dpi=150)
    print("Ablation plot → ablation_v4.png")

    return ar_ppls, jepa_guided, jepa_base


# ─────────────────────────────────────────────────────────────────────────────
# 11. TRAINING PLOTS
# ─────────────────────────────────────────────────────────────────────────────

def plot_training(history):
    fig, axes = plt.subplots(2, 3, figsize=(16, 8))
    fig.suptitle("JEPA-Guided AR Transformer v4 — Training Curves", fontsize=13)
    s = history["step"]

    def plot(ax, tk, vk, title, ct="#4A90D9", cv="#E05C5C"):
        ax.plot(s, history[tk], label="Train", color=ct, lw=1.5)
        ax.plot(s, history[vk], label="Val",   color=cv, lw=1.5)
        ax.set_title(title); ax.set_xlabel("Step")
        ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    plot(axes[0,0], "train_ppl",   "val_ppl",   "Perplexity (guided) ↓")
    plot(axes[0,1], "train_ar",    "val_ar",    "AR Loss ↓")
    plot(axes[0,2], "train_align", "val_align", "JEPA Align Loss ↓",
         ct="#8E44AD", cv="#6C3483")
    plot(axes[1,0], "train_vc",    "val_vc",    "VICReg Loss ↓",
         ct="#E67E22", cv="#CA6F1E")

    # guided vs base PPL on val
    ax = axes[1, 1]
    ax.plot(s, history["val_ppl_guided"], label="Val PPL (guided)", color="#4A90D9", lw=2)
    ax.plot(s, history["val_ppl_base"],   label="Val PPL (no guid)", color="#85C1E9",
            lw=1.5, linestyle="--")
    ax.set_title("Guidance Effect on Val PPL\n(guided vs no guidance, same model)")
    ax.set_xlabel("Step"); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    # delta plot
    ax = axes[1, 2]
    deltas = [b - g for b, g in
              zip(history["val_ppl_base"], history["val_ppl_guided"])]
    colors = ["#4A90D9" if d > 0 else "#E05C5C" for d in deltas]
    ax.bar(range(len(deltas)), deltas, color=colors)
    ax.axhline(0, color="black", lw=0.8)
    ax.set_title("Guidance delta per eval step\n(base − guided, +ve = guidance helps)")
    ax.set_xlabel("Eval step"); ax.grid(True, alpha=0.3, axis="y")

    plt.tight_layout()
    plt.savefig("training_v4.png", dpi=150)
    print("Training plot → training_v4.png")
    plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# 12. FINAL GENERATION + SAVE
# ─────────────────────────────────────────────────────────────────────────────

def final_generation(cfg, tokenizer, history):
    model = JEPAGuidedTransformer(cfg).to(cfg.device)
    model.load_state_dict(torch.load("best_model_v4.pt", map_location=cfg.device))
    model.eval()

    print("\n── Final Generation: With vs Without JEPA Guidance ────────")
    results = []
    for prompt in cfg.gen_prompts:
        pid       = torch.tensor([tokenizer.encode(prompt).ids], dtype=torch.long)
        with_g    = decode(model.generate(pid, max_new=80, use_guidance=True),  tokenizer)
        without_g = decode(model.generate(pid, max_new=80, use_guidance=False), tokenizer)
        print(f"\n  Prompt : \"{prompt}\"")
        print(f"  [WITH]   {with_g}")
        print(f"  [WITHOUT]{without_g}")
        results.append({"prompt": prompt,
                        "with_guidance": with_g,
                        "without_guidance": without_g})

    out = {
        "version": "v4",
        "key_change": "JEPA guidance injected via cross-attention into representation "
                      "space (before ar_head), not logit space",
        "best_val_ppl_guided": min(history["val_ppl_guided"]),
        "final_val_ppl_guided": history["val_ppl_guided"][-1],
        "final_val_ppl_base":   history["val_ppl_base"][-1],
        "final_guidance_delta": history["val_ppl_base"][-1] - history["val_ppl_guided"][-1],
        "final_align": history["val_align"][-1],
        "final_vc":    history["val_vc"][-1],
        "generation_samples": results,
        "training_history": {
            "steps":          history["step"],
            "val_ppl_guided": history["val_ppl_guided"],
            "val_ppl_base":   history["val_ppl_base"],
            "val_align":      history["val_align"],
            "val_vc":         history["val_vc"],
        },
        "config": {k: v for k, v in cfg.__dict__.items() if not callable(v)},
    }
    with open("jepa_ar_v4_results.json", "w") as f:
        json.dump(out, f, indent=2)
    print("\nResults saved → jepa_ar_v4_results.json")


# ─────────────────────────────────────────────────────────────────────────────
# 13. ENTRY POINT
# ─────────────────────────────────────────────────────────────────────────────

if __name__ == "__main__":
    t0 = time.time()

    model, tokenizer, history = train(CFG)
    plot_training(history)
    final_generation(CFG, tokenizer, history)

    print("\n── Re-loading data for ablation ────────────────────────────")
    train_text, val_text = load_wikitext2()
    train_chunks = tokenize_and_chunk(train_text, tokenizer, CFG.seq_len)
    val_chunks   = tokenize_and_chunk(val_text,   tokenizer, CFG.seq_len)
    train_loader = DataLoader(PairDataset(train_chunks),
                              batch_size=CFG.batch_size, shuffle=True)
    val_loader   = DataLoader(PairDataset(val_chunks),
                              batch_size=CFG.batch_size, shuffle=False)
    run_ablation(train_loader, val_loader, tokenizer, CFG, epochs=6)

    print(f"\nTotal time: {(time.time()-t0)/60:.1f} min")
    print("\nOutput files:")
    print("  best_model_v4.pt")
    print("  training_v4.png")
    print("  ablation_v4.png")
    print("  jepa_ar_v4_results.json")
```

```python
"""
JEPA-Guided AR Transformer  —  v5
===================================
Upgrades from v4:
  1. WikiText-103 dataset  (~500MB, 103M tokens vs 2MB)
     → much richer training signal, more realistic test of the idea
  2. Larger model: d_model=256, n_layers=6, n_heads=8, d_ff=1024
     → ~13M params vs ~1.9M, tests whether guidance delta scales with model size
  3. More training: 15 epochs with cosine LR decay
     → WikiText-103 has 60x more data, needs more steps to converge
  4. Repetition fix: guidance_alpha annealed during inference
     → starts at 0.4 (strong early signal) decays to 0.1 (avoids repetition loops)
  5. Checkpoint saving every N steps + resume support
     → long runs need resumability on Colab

Architecture (unchanged from v4 — validated):
  backbone → h
  h.detach() → jepa_proj → jepa_pred          (JEPA loss, detached)
  jepa_pred.mean().detach() → GuidanceCrossAttn(h, ·) → correction
  h + correction → ar_head → ar_logits        (AR loss end-to-end)

Hardware targets:
  Free Colab T4  (16GB): batch=32, seq=128  → ~45-60 min
  Colab Pro A100 (40GB): batch=64, seq=256  → ~40-50 min   [set COLAB_PRO=True below]

Install:
    pip install torch datasets tokenizers matplotlib tqdm

Run:
    python jepa_ar_v5.py
"""

import os, math, time, json, random
from copy import deepcopy
from dataclasses import dataclass, field
from pathlib import Path

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt
from tqdm import tqdm

# ── reproducibility ───────────────────────────────────────────────────────────
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

# ─────────────────────────────────────────────────────────────────────────────
# HARDWARE FLAG  —  set True if you have Colab Pro / A100
# ─────────────────────────────────────────────────────────────────────────────
COLAB_PRO = False   # ← change to True for A100

# ─────────────────────────────────────────────────────────────────────────────
# 1.  CONFIG
# ─────────────────────────────────────────────────────────────────────────────
@dataclass
class Config:
    # ── model (scaled up from v4) ─────────────────────────────────────────
    vocab_size        : int   = 16000        # larger vocab for bigger corpus
    seq_len           : int   = 256 if COLAB_PRO else 128
    d_model           : int   = 256          # was 128
    n_heads           : int   = 8            # was 4
    n_layers          : int   = 6            # was 4
    d_ff              : int   = 1024         # was 512
    dropout           : float = 0.1

    # ── JEPA ─────────────────────────────────────────────────────────────
    jepa_dim          : int   = 128          # was 64, scaled with d_model
    jepa_ema_decay    : float = 0.996
    jepa_lambda       : float = 1.0

    # ── VICReg ───────────────────────────────────────────────────────────
    vicreg_mu         : float = 1.0
    vicreg_nu         : float = 0.04
    vicreg_eps        : float = 1e-4

    # ── guidance cross-attention ──────────────────────────────────────────
    guidance_n_heads  : int   = 4            # was 2, scaled with d_model
    guidance_init_scale : float = 0.0        # zero-init → starts neutral

    # ── inference repetition fix ──────────────────────────────────────────
    # guidance_alpha is annealed during generation:
    #   first tokens: alpha_max (strong future signal)
    #   later tokens: alpha_min (avoid repetition loops)
    guidance_alpha_max : float = 0.4
    guidance_alpha_min : float = 0.1
    guidance_anneal_steps : int = 20         # tokens over which to anneal

    # ── training ─────────────────────────────────────────────────────────
    batch_size        : int   = 64 if COLAB_PRO else 32
    lr                : float = 2e-4         # slightly lower for larger model
    epochs            : int   = 7 #10 #15           # more epochs for larger dataset
    grad_clip         : float = 1.0
    warmup_steps      : int   = 1000         # longer warmup for larger model
    lr_min            : float = 2e-5         # cosine decay floor

    # ── checkpointing ─────────────────────────────────────────────────────
    save_every        : int   = 1000         # save checkpoint every N steps
    checkpoint_dir    : str   = "checkpoints_v5"
    resume            : bool  = True         # auto-resume from latest checkpoint

    # ── eval / logging ────────────────────────────────────────────────────
    eval_every        : int   = 500
    gen_every         : int   = 1000
    gen_prompts       : list  = field(default_factory=lambda: [
        "The history of science",
        "In the early 20th century",
        "The political system of",
        "During the Second World",
    ])
    device : str = "cuda" if torch.cuda.is_available() else "cpu"

CFG = Config()
print(f"Device  : {CFG.device}")
print(f"d_model : {CFG.d_model}  n_layers: {CFG.n_layers}  "
      f"seq_len: {CFG.seq_len}  batch: {CFG.batch_size}")
print(f"Dataset : WikiText-103")
print(f"Epochs  : {CFG.epochs}")


# ─────────────────────────────────────────────────────────────────────────────
# 2.  DATA  —  WikiText-103
# ─────────────────────────────────────────────────────────────────────────────

def load_wikitext103():
    from datasets import load_dataset
    print("  Downloading WikiText-103 (~500MB, first run only)...")
    ds = load_dataset("wikitext", "wikitext-103-raw-v1")
    # Return the HuggingFace dataset split directly — we'll tokenise
    # row-by-row below instead of joining into one giant string
    return ds["train"], ds["validation"]


def build_tokenizer(train_ds, vocab_size=16000):
    from tokenizers import Tokenizer
    from tokenizers.models import BPE
    from tokenizers.trainers import BpeTrainer
    from tokenizers.pre_tokenizers import Whitespace
    import tempfile
    print(f"  Training BPE tokenizer on first 50K rows (vocab={vocab_size})...")
    tok = Tokenizer(BPE(unk_token="[UNK]"))
    tok.pre_tokenizer = Whitespace()
    trainer = BpeTrainer(vocab_size=vocab_size,
                         special_tokens=["[PAD]","[UNK]","[BOS]","[EOS]"])
    # write a sample to a temp file — fast, avoids joining 540MB
    sample_rows = min(50_000, len(train_ds))
    with tempfile.NamedTemporaryFile("w", suffix=".txt", delete=False,
                                     encoding="utf-8") as f:
        for row in train_ds.select(range(sample_rows)):
            line = row["text"].strip()
            if line:
                f.write(line + "\n")
        tmp = f.name
    tok.train([tmp], trainer)
    os.unlink(tmp)
    return tok


def tokenize_and_chunk(ds_split, tokenizer, seq_len, desc="",
                        max_rows=None):
    """
    Tokenise a HuggingFace dataset split row-by-row and chunk into
    fixed-length sequences.  Much faster than joining into one string
    and avoids OOM on large corpora.
    """
    print(f"  Tokenising {desc} (row-by-row)...")
    all_ids = []
    rows = ds_split if max_rows is None else ds_split.select(range(max_rows))
    for row in tqdm(rows, desc=f"  tok {desc}", leave=False):
        text = row["text"].strip()
        if not text:
            continue
        all_ids.extend(tokenizer.encode(text).ids)

    print(f"    {desc}: {len(all_ids):,} tokens")
    n = (len(all_ids) // seq_len) * seq_len
    return np.array(all_ids[:n], dtype=np.int32).reshape(-1, seq_len)


class PairDataset(Dataset):
    """Returns (ctx_chunk[i], future_chunk[i+1]) pairs."""
    def __init__(self, chunks):
        self.chunks = torch.tensor(chunks, dtype=torch.long)

    def __len__(self):
        return len(self.chunks) - 1

    def __getitem__(self, i):
        return self.chunks[i], self.chunks[i + 1]


# ─────────────────────────────────────────────────────────────────────────────
# 3.  MODEL COMPONENTS
# ─────────────────────────────────────────────────────────────────────────────

class RotaryEmbedding(nn.Module):
    def __init__(self, dim):
        super().__init__()
        inv_freq = 1.0 / (10000 ** (torch.arange(0, dim, 2).float() / dim))
        self.register_buffer("inv_freq", inv_freq)

    def forward(self, seq_len, device):
        t     = torch.arange(seq_len, device=device).float()
        freqs = torch.outer(t, self.inv_freq)
        emb   = torch.cat([freqs, freqs], dim=-1)
        return emb.cos(), emb.sin()


def rotate_half(x):
    x1, x2 = x.chunk(2, dim=-1)
    return torch.cat([-x2, x1], dim=-1)


def apply_rope(q, k, cos, sin):
    return q * cos + rotate_half(q) * sin, k * cos + rotate_half(k) * sin


class MultiHeadSelfAttention(nn.Module):
    def __init__(self, d_model, n_heads, dropout, causal=True):
        super().__init__()
        self.n_heads = n_heads
        self.d_head  = d_model // n_heads
        self.causal  = causal
        self.qkv     = nn.Linear(d_model, 3 * d_model, bias=False)
        self.out     = nn.Linear(d_model, d_model, bias=False)
        self.drop    = nn.Dropout(dropout)
        self.rope    = RotaryEmbedding(self.d_head)

    def forward(self, x):
        B, T, C = x.shape
        cos, sin = self.rope(T, x.device)
        cos = cos.unsqueeze(0).unsqueeze(0)
        sin = sin.unsqueeze(0).unsqueeze(0)
        qkv = self.qkv(x).reshape(B, T, 3, self.n_heads, self.d_head)
        q, k, v = qkv.unbind(2)
        q = q.transpose(1, 2); k = k.transpose(1, 2); v = v.transpose(1, 2)
        q, k = apply_rope(q, k, cos, sin)
        scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(self.d_head)
        if self.causal:
            mask = torch.triu(torch.ones(T, T, device=x.device), diagonal=1).bool()
            scores = scores.masked_fill(mask, float("-inf"))
        attn = self.drop(F.softmax(scores, dim=-1))
        out  = torch.matmul(attn, v).transpose(1, 2).reshape(B, T, C)
        return self.out(out)


class FFN(nn.Module):
    def __init__(self, d_model, d_ff, dropout):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_model, d_ff), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(d_ff, d_model), nn.Dropout(dropout))

    def forward(self, x):
        return self.net(x)


class TransformerBlock(nn.Module):
    def __init__(self, d_model, n_heads, d_ff, dropout):
        super().__init__()
        self.attn = MultiHeadSelfAttention(d_model, n_heads, dropout)
        self.ffn  = FFN(d_model, d_ff, dropout)
        self.ln1  = nn.LayerNorm(d_model)
        self.ln2  = nn.LayerNorm(d_model)

    def forward(self, x):
        x = x + self.attn(self.ln1(x))
        x = x + self.ffn(self.ln2(x))
        return x


# ─────────────────────────────────────────────────────────────────────────────
# 4.  GUIDANCE CROSS-ATTENTION
# ─────────────────────────────────────────────────────────────────────────────

class GuidanceCrossAttention(nn.Module):
    """
    Lightweight cross-attention injecting JEPA future context into backbone repr.
    queries  = backbone h         (B, T, d_model)
    keys/val = jepa_summary       (B, 1, d_model)   ← mean-pooled future embedding
    output   = residual correction to h, zero-initialised → starts as identity
    """
    def __init__(self, d_model, jepa_dim, n_heads, dropout, init_scale=0.0):
        super().__init__()
        assert d_model % n_heads == 0
        self.n_heads = n_heads
        self.d_head  = d_model // n_heads
        self.scale   = math.sqrt(self.d_head)
        self.jepa_to_kv = nn.Linear(jepa_dim, 2 * d_model, bias=False)
        self.q_proj     = nn.Linear(d_model,  d_model,     bias=False)
        self.out_proj   = nn.Linear(d_model,  d_model,     bias=False)
        nn.init.constant_(self.out_proj.weight, init_scale)
        self.ln   = nn.LayerNorm(d_model)
        self.drop = nn.Dropout(dropout)

    def forward(self, h, jepa_summary, alpha=1.0):
        """
        h            : (B, T, d_model)
        jepa_summary : (B, jepa_dim)    mean-pooled JEPA embedding
        alpha        : scalar gate — set to 0 to disable guidance entirely
        """
        if alpha == 0.0:
            return torch.zeros_like(h)

        B, T, C = h.shape
        q  = self.q_proj(self.ln(h))
        q  = q.reshape(B, T, self.n_heads, self.d_head).transpose(1, 2)

        kv = self.jepa_to_kv(jepa_summary).unsqueeze(1)   # (B, 1, 2*C)
        k, v = kv.chunk(2, dim=-1)
        k  = k.reshape(B, 1, self.n_heads, self.d_head).transpose(1, 2)
        v  = v.reshape(B, 1, self.n_heads, self.d_head).transpose(1, 2)

        scores = torch.matmul(q, k.transpose(-2, -1)) / self.scale  # (B,H,T,1)
        attn   = self.drop(F.softmax(scores, dim=-1))
        out    = torch.matmul(attn, v).transpose(1, 2).reshape(B, T, C)
        return alpha * self.out_proj(out)


# ─────────────────────────────────────────────────────────────────────────────
# 5.  VICReg
# ─────────────────────────────────────────────────────────────────────────────

def vicreg_loss(z, cfg):
    if z.dim() == 3:
        B, T, D = z.shape; z = z.reshape(B * T, D)
    z = z - z.mean(dim=0)
    N, D = z.shape
    std      = torch.sqrt(z.var(dim=0) + cfg.vicreg_eps)
    var_loss = torch.relu(1.0 - std).mean()
    cov      = (z.T @ z) / (N - 1)
    off_diag = cov * (1 - torch.eye(D, device=z.device))
    cov_loss = off_diag.pow(2).sum() / D
    return cfg.vicreg_mu * var_loss + cfg.vicreg_nu * cov_loss


# ─────────────────────────────────────────────────────────────────────────────
# 6.  FULL MODEL
# ─────────────────────────────────────────────────────────────────────────────

class JEPAGuidedTransformer(nn.Module):
    def __init__(self, cfg: Config):
        super().__init__()
        self.cfg = cfg

        # backbone
        self.embed    = nn.Embedding(cfg.vocab_size, cfg.d_model)
        self.drop     = nn.Dropout(cfg.dropout)
        self.blocks   = nn.ModuleList([
            TransformerBlock(cfg.d_model, cfg.n_heads, cfg.d_ff, cfg.dropout)
            for _ in range(cfg.n_layers)
        ])
        self.ln_final = nn.LayerNorm(cfg.d_model)

        # AR head
        self.ar_head  = nn.Linear(cfg.d_model, cfg.vocab_size, bias=False)
        self.ar_head.weight = self.embed.weight

        # JEPA online projection (backbone always detached)
        self.jepa_proj = nn.Sequential(
            nn.Linear(cfg.d_model, cfg.d_model),
            nn.GELU(),
            nn.Linear(cfg.d_model, cfg.jepa_dim),
        )

        # JEPA target encoder (EMA)
        self.target_embed  = deepcopy(self.embed)
        self.target_blocks = nn.ModuleList([deepcopy(b) for b in self.blocks])
        self.target_ln     = deepcopy(self.ln_final)
        self.target_proj   = deepcopy(self.jepa_proj)
        for p in self._target_params():
            p.requires_grad_(False)

        # Guidance cross-attention (sits between backbone and ar_head)
        self.guidance_xattn = GuidanceCrossAttention(
            d_model    = cfg.d_model,
            jepa_dim   = cfg.jepa_dim,
            n_heads    = cfg.guidance_n_heads,
            dropout    = cfg.dropout,
            init_scale = cfg.guidance_init_scale,
        )

        self._init_weights()

    def _target_params(self):
        for m in [self.target_embed, self.target_ln,
                  self.target_proj, *self.target_blocks]:
            yield from m.parameters()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, std=0.02)
                if m.bias is not None: nn.init.zeros_(m.bias)
            elif isinstance(m, nn.Embedding):
                nn.init.normal_(m.weight, std=0.02)
        nn.init.constant_(self.guidance_xattn.out_proj.weight,
                          self.cfg.guidance_init_scale)

    @torch.no_grad()
    def update_ema(self):
        d = self.cfg.jepa_ema_decay
        for online, target in [
            (self.embed,     self.target_embed),
            (self.ln_final,  self.target_ln),
            (self.jepa_proj, self.target_proj),
        ]:
            for po, pt in zip(online.parameters(), target.parameters()):
                pt.data.mul_(d).add_(po.data, alpha=1 - d)
        for bo, bt in zip(self.blocks, self.target_blocks):
            for po, pt in zip(bo.parameters(), bt.parameters()):
                pt.data.mul_(d).add_(po.data, alpha=1 - d)

    def _encode_backbone(self, x):
        h = self.drop(self.embed(x))
        for b in self.blocks: h = b(h)
        return self.ln_final(h)

    @torch.no_grad()
    def _encode_target(self, x):
        h = self.target_embed(x)
        for b in self.target_blocks: h = b(h)
        return self.target_ln(h)

    def forward(self, ctx, future=None, use_guidance=True, guidance_alpha=None):
        h         = self._encode_backbone(ctx)
        jepa_pred = self.jepa_proj(h.detach())          # backbone detached

        jepa_target = None
        if future is not None:
            with torch.no_grad():
                h_fut       = self._encode_target(future)
                jepa_target = self.target_proj(h_fut)

        if use_guidance:
            alpha        = guidance_alpha if guidance_alpha is not None \
                           else self.cfg.guidance_alpha_max
            jepa_summary = jepa_pred.mean(dim=1).detach()
            correction   = self.guidance_xattn(h, jepa_summary, alpha=alpha)
            h_guided     = h + correction
        else:
            h_guided = h

        ar_logits = self.ar_head(h_guided)
        return ar_logits, jepa_pred, jepa_target

    @torch.no_grad()
    def generate(self, prompt_ids, max_new=100, temperature=0.85,
                 top_k=50, top_p=0.92, use_guidance=True):
        """
        Generation with:
          - top-k + nucleus (top-p) sampling  → better diversity than top-k alone
          - annealed guidance alpha            → strong early, weak late (fixes repetition)
          - repetition penalty                 → discourages exact token repeats
        """
        self.eval()
        device = next(self.parameters()).device
        ids    = prompt_ids.to(device)
        cfg    = self.cfg

        for step in range(max_new):
            ctx = ids[:, -cfg.seq_len:]

            # anneal guidance alpha: strong at start, taper to min
            if use_guidance:
                t     = min(step / max(cfg.guidance_anneal_steps, 1), 1.0)
                alpha = cfg.guidance_alpha_max * (1 - t) + cfg.guidance_alpha_min * t
            else:
                alpha = 0.0

            ar_logits, _, _ = self.forward(ctx, future=None,
                                           use_guidance=use_guidance,
                                           guidance_alpha=alpha)
            logits = ar_logits[:, -1, :] / temperature

            # repetition penalty — discount tokens already generated
            if ids.shape[1] > 1:
                for token_id in ids[0, -50:].unique():
                    logits[0, token_id] /= 1.3

            # top-k filter
            if top_k > 0:
                topk_v, _ = torch.topk(logits, min(top_k, logits.size(-1)))
                logits[logits < topk_v[:, -1:]] = float("-inf")

            # nucleus (top-p) filter
            if top_p < 1.0:
                sorted_logits, sorted_idx = torch.sort(logits, descending=True)
                cum_probs = torch.cumsum(F.softmax(sorted_logits, dim=-1), dim=-1)
                remove    = cum_probs - F.softmax(sorted_logits, dim=-1) > top_p
                sorted_logits[remove] = float("-inf")
                logits = torch.zeros_like(logits).scatter_(
                    1, sorted_idx, sorted_logits)

            ids = torch.cat([ids,
                             torch.multinomial(F.softmax(logits, -1), 1)], dim=1)

        self.train()
        return ids[0].tolist()


# ─────────────────────────────────────────────────────────────────────────────
# 7.  LOSS
# ─────────────────────────────────────────────────────────────────────────────

def compute_loss(ar_logits, jepa_pred, jepa_target, ctx_tokens, cfg):
    B, T, V = ar_logits.shape

    ar_loss = F.cross_entropy(
        ar_logits[:, :-1, :].reshape(-1, V),
        ctx_tokens[:, 1:].reshape(-1),
        ignore_index=0,
    )

    jepa_loss = torch.tensor(0.0, device=ar_logits.device)
    align_val = vc_val = 0.0
    if jepa_target is not None:
        pred_vec   = F.normalize(jepa_pred[:, -1, :],     dim=-1)
        target_vec = F.normalize(jepa_target.mean(dim=1), dim=-1)
        align_loss = F.smooth_l1_loss(pred_vec, target_vec)
        vc_loss    = vicreg_loss(jepa_pred, cfg)
        jepa_loss  = align_loss + vc_loss
        align_val  = align_loss.item()
        vc_val     = vc_loss.item()

    total = ar_loss + cfg.jepa_lambda * jepa_loss
    return total, ar_loss.item(), align_val, vc_val


# ─────────────────────────────────────────────────────────────────────────────
# 8.  LR SCHEDULE  —  linear warmup + cosine decay
# ─────────────────────────────────────────────────────────────────────────────

def get_lr(step, total_steps, cfg):
    if step < cfg.warmup_steps:
        return cfg.lr * step / max(1, cfg.warmup_steps)
    # cosine decay from lr → lr_min
    progress = (step - cfg.warmup_steps) / max(1, total_steps - cfg.warmup_steps)
    cosine   = 0.5 * (1 + math.cos(math.pi * progress))
    return cfg.lr_min + (cfg.lr - cfg.lr_min) * cosine


# ─────────────────────────────────────────────────────────────────────────────
# 9.  CHECKPOINT UTILS
# ─────────────────────────────────────────────────────────────────────────────

def save_checkpoint(model, optimizer, step, history, cfg):
    Path(cfg.checkpoint_dir).mkdir(exist_ok=True)
    path = os.path.join(cfg.checkpoint_dir, f"ckpt_step{step}.pt")
    torch.save({
        "step":      step,
        "model":     model.state_dict(),
        "optimizer": optimizer.state_dict(),
        "history":   history,
    }, path)
    # keep only last 3 checkpoints to save disk space
    ckpts = sorted(Path(cfg.checkpoint_dir).glob("ckpt_step*.pt"),
                   key=lambda p: int(p.stem.split("step")[1]))
    for old in ckpts[:-3]:
        old.unlink()
    return path


def load_latest_checkpoint(model, optimizer, cfg):
    ckpts = sorted(Path(cfg.checkpoint_dir).glob("ckpt_step*.pt"),
                   key=lambda p: int(p.stem.split("step")[1]))
    if not ckpts:
        return 0, None
    path = ckpts[-1]
    print(f"  Resuming from {path}")
    ckpt = torch.load(path, map_location=cfg.device)
    model.load_state_dict(ckpt["model"])
    optimizer.load_state_dict(ckpt["optimizer"])
    return ckpt["step"], ckpt["history"]


# ─────────────────────────────────────────────────────────────────────────────
# 10. EVAL
# ─────────────────────────────────────────────────────────────────────────────

@torch.no_grad()
def evaluate(model, loader, cfg, max_batches=100):
    model.eval()
    tot = dict(ar=0, align=0, vc=0, n=0)
    for i, (ctx, future) in enumerate(loader):
        if i >= max_batches: break
        ctx, future = ctx.to(cfg.device), future.to(cfg.device)
        ar_logits, jp, jt = model(ctx, future, use_guidance=True)
        _, ar_l, al_l, vc_l = compute_loss(ar_logits, jp, jt, ctx, cfg)
        tot["ar"] += ar_l; tot["align"] += al_l; tot["vc"] += vc_l; tot["n"] += 1
    model.train()
    n = tot["n"]
    return tot["ar"]/n, tot["align"]/n, tot["vc"]/n


@torch.no_grad()
def evaluate_both(model, loader, cfg, max_batches=100):
    """Returns (ppl_guided, ppl_base) — key metric for ablation."""
    model.eval()
    ar_g = ar_b = n = 0
    V = cfg.vocab_size
    for i, (ctx, future) in enumerate(loader):
        if i >= max_batches: break
        ctx, future = ctx.to(cfg.device), future.to(cfg.device)
        lg, _, _ = model(ctx, future, use_guidance=True)
        lb, _, _ = model(ctx, future, use_guidance=False)
        ar_g += F.cross_entropy(lg[:,:-1].reshape(-1,V),
                                ctx[:,1:].reshape(-1), ignore_index=0).item()
        ar_b += F.cross_entropy(lb[:,:-1].reshape(-1,V),
                                ctx[:,1:].reshape(-1), ignore_index=0).item()
        n += 1
    model.train()
    return math.exp(ar_g/n), math.exp(ar_b/n)


def decode(ids, tokenizer):
    return tokenizer.decode(ids)


def print_generation(model, tokenizer, cfg, step):
    print(f"\n  ── Generation at step {step} ──")
    for prompt in cfg.gen_prompts[:2]:
        pid = torch.tensor([tokenizer.encode(prompt).ids], dtype=torch.long)
        with_g    = decode(model.generate(pid, max_new=80, use_guidance=True),  tokenizer)
        without_g = decode(model.generate(pid, max_new=80, use_guidance=False), tokenizer)
        print(f"\n  Prompt : \"{prompt}\"")
        print(f"  [WITH]   {with_g}")
        print(f"  [WITHOUT]{without_g}")


# ─────────────────────────────────────────────────────────────────────────────
# 11. TRAINING LOOP
# ─────────────────────────────────────────────────────────────────────────────

def train(cfg, train_loader, val_loader, tokenizer):
    print("\n── Building model ─────────────────────────────────────────")
    model = JEPAGuidedTransformer(cfg).to(cfg.device)
    n_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"  Trainable parameters: {n_params:,}")

    optimizer = torch.optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=cfg.lr, weight_decay=0.01, betas=(0.9, 0.95))

    # ── resume from checkpoint if available ──────────────────────────────
    start_step = 0
    history    = dict(step=[], train_ar=[], train_align=[], train_vc=[],
                      val_ppl_guided=[], val_ppl_base=[],
                      train_ppl=[], val_ppl=[],
                      val_align=[], val_vc=[], lr=[])

    if cfg.resume:
        start_step, saved_history = load_latest_checkpoint(model, optimizer, cfg)
        if saved_history:
            history = saved_history
            print(f"  Resumed at step {start_step}, "
                  f"best val PPL so far: {min(history['val_ppl_guided']):.1f}")

    # estimate total steps for LR schedule
    steps_per_epoch = len(train_loader)
    total_steps     = steps_per_epoch * cfg.epochs
    print(f"  Steps per epoch : {steps_per_epoch:,}")
    print(f"  Total steps     : {total_steps:,}")

    step         = start_step
    best_val_ppl = min(history["val_ppl_guided"]) if history["val_ppl_guided"] \
                   else float("inf")

    print("\n── Training ────────────────────────────────────────────────")

    for epoch in range(1, cfg.epochs + 1):
        ep_ar = ep_align = ep_vc = ep_n = 0
        pbar  = tqdm(train_loader, desc=f"Epoch {epoch}/{cfg.epochs}")

        for ctx, future in pbar:
            # skip steps already done (resume)
            if step < start_step:
                step += 1; continue

            ctx, future = ctx.to(cfg.device), future.to(cfg.device)
            lr = get_lr(step, total_steps, cfg)
            for pg in optimizer.param_groups: pg["lr"] = lr

            optimizer.zero_grad()
            ar_logits, jp, jt = model(ctx, future, use_guidance=True)
            loss, ar_l, al_l, vc_l = compute_loss(ar_logits, jp, jt, ctx, cfg)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.grad_clip)
            optimizer.step()
            model.update_ema()

            ep_ar += ar_l; ep_align += al_l; ep_vc += vc_l
            ep_n  += 1; step += 1

            pbar.set_postfix({
                "ar":    f"{ar_l:.3f}",
                "align": f"{al_l:.4f}",
                "ppl":   f"{math.exp(ar_l):.1f}",
                "lr":    f"{lr:.2e}",
            })

            # ── eval ──────────────────────────────────────────────────────
            if step % cfg.eval_every == 0:
                val_ar, val_al, val_vc = evaluate(model, val_loader, cfg)
                ppl_guided, ppl_base   = evaluate_both(model, val_loader, cfg)
                train_ppl = math.exp(ep_ar / ep_n)

                history["step"].append(step)
                history["train_ar"].append(ep_ar / ep_n)
                history["train_align"].append(ep_align / ep_n)
                history["train_vc"].append(ep_vc / ep_n)
                history["val_ppl_guided"].append(ppl_guided)
                history["val_ppl_base"].append(ppl_base)
                history["train_ppl"].append(train_ppl)
                history["val_ppl"].append(ppl_guided)
                history["val_align"].append(val_al)
                history["val_vc"].append(val_vc)
                history["lr"].append(lr)

                delta = ppl_base - ppl_guided
                if ppl_guided < best_val_ppl:
                    best_val_ppl = ppl_guided
                    torch.save(model.state_dict(), "best_model_v5.pt")

                print(f"\n  Step {step:6d} | "
                      f"Train PPL {train_ppl:.1f} | "
                      f"Val guided {ppl_guided:.1f} | "
                      f"base {ppl_base:.1f} | "
                      f"Δ {delta:+.2f} | "
                      f"align {val_al:.4f} | "
                      f"lr {lr:.2e}")

            # ── generation ────────────────────────────────────────────────
            if step % cfg.gen_every == 0:
                print_generation(model, tokenizer, cfg, step)

            # ── checkpoint ────────────────────────────────────────────────
            if step % cfg.save_every == 0:
                path = save_checkpoint(model, optimizer, step, history, cfg)
                print(f"  Checkpoint → {path}")

    # final save
    torch.save(model.state_dict(), "best_model_v5.pt")
    return model, history


# ─────────────────────────────────────────────────────────────────────────────
# 12. ABLATION  —  AR-only vs JEPA-guided (6 epochs on WikiText-103 subset)
# ─────────────────────────────────────────────────────────────────────────────

def run_ablation(train_loader, val_loader, cfg, epochs=6):
    """
    Use a subset of WikiText-103 for ablation to keep runtime reasonable.
    Takes first 20% of train_loader batches per epoch.
    """
    print(f"\n── Ablation: AR-only vs JEPA-guided v5 ({epochs} epochs) ───")
    MAX_BATCHES_PER_EPOCH = len(train_loader) // 5   # 20% of data

    def train_variant(use_jepa, label):
        cfg2 = deepcopy(cfg)
        if not use_jepa:
            cfg2.jepa_lambda = 0.0

        m   = JEPAGuidedTransformer(cfg2).to(cfg.device)
        opt = torch.optim.AdamW(
            [p for p in m.parameters() if p.requires_grad],
            lr=cfg2.lr, weight_decay=0.01, betas=(0.9, 0.95))

        ppls_guided = []; ppls_base = []

        for ep in range(epochs):
            batch_count = 0
            for ctx, future in tqdm(train_loader,
                                    desc=f"  {label} ep{ep+1}",
                                    leave=False,
                                    total=MAX_BATCHES_PER_EPOCH):
                if batch_count >= MAX_BATCHES_PER_EPOCH: break
                ctx, future = ctx.to(cfg.device), future.to(cfg.device)
                opt.zero_grad()
                ar_logits, jp, jt = m(ctx, future,
                                      use_guidance=use_jepa)
                loss, _, _, _ = compute_loss(ar_logits, jp, jt, ctx, cfg2)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(m.parameters(), cfg2.grad_clip)
                opt.step(); m.update_ema()
                batch_count += 1

            if use_jepa:
                pg, pb = evaluate_both(m, val_loader, cfg2)
            else:
                pb, _ = evaluate_both(m, val_loader, cfg2)
                pg = pb
            ppls_guided.append(pg); ppls_base.append(pb)
            print(f"    {label}  ep{ep+1}  guided={pg:.2f}  base={pb:.2f}  "
                  f"Δ={pb-pg:+.2f}")
        return ppls_guided, ppls_base

    ar_ppls, _             = train_variant(False, "AR-only  ")
    jepa_guided, jepa_base = train_variant(True,  "JEPA+AR  ")

    delta_guidance = jepa_base[-1] - jepa_guided[-1]
    delta_vs_ar    = ar_ppls[-1]   - jepa_guided[-1]

    print(f"\n  ── Final (epoch {epochs}) ──")
    print(f"  AR-only              : {ar_ppls[-1]:.2f}")
    print(f"  JEPA no guidance     : {jepa_base[-1]:.2f}")
    print(f"  JEPA guided          : {jepa_guided[-1]:.2f}")
    print(f"  Guidance benefit     : {delta_guidance:+.2f}  "
          f"({'✓' if delta_guidance > 0 else '✗'})")
    print(f"  vs AR-only           : {delta_vs_ar:+.2f}  "
          f"({'✓ JEPA beats AR' if delta_vs_ar > 0 else '✗ AR still better'})")

    ep_range = list(range(1, epochs + 1))
    fig, axes = plt.subplots(1, 3, figsize=(16, 4))
    fig.suptitle("v5 Ablation: AR-only vs JEPA-Guided (WikiText-103 subset)", fontsize=12)

    axes[0].plot(ep_range, ar_ppls,     "o-",  label="AR only",          color="#E05C5C", lw=2)
    axes[0].plot(ep_range, jepa_guided, "s-",  label="JEPA guided",      color="#4A90D9", lw=2)
    axes[0].plot(ep_range, jepa_base,   "^--", label="JEPA no guidance", color="#85C1E9", lw=1.5)
    axes[0].set_title("Val PPL ↓"); axes[0].set_xlabel("Epoch")
    axes[0].legend(fontsize=8); axes[0].grid(True, alpha=0.3)

    deltas_g = [b - g for b, g in zip(jepa_base, jepa_guided)]
    c1 = ["#4A90D9" if d > 0 else "#E05C5C" for d in deltas_g]
    axes[1].bar(ep_range, deltas_g, color=c1)
    axes[1].axhline(0, color="black", lw=0.8)
    axes[1].set_title("Guidance benefit (base−guided) ↑")
    axes[1].set_xlabel("Epoch"); axes[1].grid(True, alpha=0.3, axis="y")

    deltas_ar = [ar - jg for ar, jg in zip(ar_ppls, jepa_guided)]
    c2 = ["#4A90D9" if d > 0 else "#E05C5C" for d in deltas_ar]
    axes[2].bar(ep_range, deltas_ar, color=c2)
    axes[2].axhline(0, color="black", lw=0.8)
    axes[2].set_title("JEPA guided vs AR-only ↑")
    axes[2].set_xlabel("Epoch"); axes[2].grid(True, alpha=0.3, axis="y")

    plt.tight_layout()
    plt.savefig("ablation_v5.png", dpi=150)
    print("Ablation plot → ablation_v5.png")
    return ar_ppls, jepa_guided, jepa_base


# ─────────────────────────────────────────────────────────────────────────────
# 13. PLOTS
# ─────────────────────────────────────────────────────────────────────────────

def plot_training(history):
    fig, axes = plt.subplots(2, 3, figsize=(16, 8))
    fig.suptitle("JEPA-Guided AR Transformer v5 — WikiText-103", fontsize=13)
    s = history["step"]

    def plot(ax, tk, vk, title, ct="#4A90D9", cv="#E05C5C"):
        ax.plot(s, history[tk], label="Train", color=ct, lw=1.5)
        ax.plot(s, history[vk], label="Val",   color=cv, lw=1.5)
        ax.set_title(title); ax.set_xlabel("Step")
        ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    plot(axes[0,0], "train_ppl",   "val_ppl",   "Perplexity (guided) ↓")
    plot(axes[0,1], "train_ar",    "train_ar",  "AR Loss ↓")
    plot(axes[0,2], "train_align", "val_align", "JEPA Align Loss ↓",
         ct="#8E44AD", cv="#6C3483")
    plot(axes[1,0], "train_vc",    "val_vc",    "VICReg Loss ↓",
         ct="#E67E22", cv="#CA6F1E")

    ax = axes[1, 1]
    ax.plot(s, history["val_ppl_guided"], label="Val PPL (guided)", color="#4A90D9", lw=2)
    ax.plot(s, history["val_ppl_base"],   label="Val PPL (base)",   color="#85C1E9",
            lw=1.5, linestyle="--")
    ax.set_title("Guidance Effect on Val PPL")
    ax.set_xlabel("Step"); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    ax = axes[1, 2]
    if history["lr"]:
        ax.plot(s, history["lr"], color="#27AE60", lw=1.5)
        ax.set_title("Learning Rate (cosine decay)")
        ax.set_xlabel("Step"); ax.grid(True, alpha=0.3)
        ax.yaxis.set_major_formatter(
            plt.FuncFormatter(lambda x, _: f"{x:.1e}"))

    plt.tight_layout()
    plt.savefig("training_v5.png", dpi=150)
    print("Training plot → training_v5.png")
    plt.show()


# ─────────────────────────────────────────────────────────────────────────────
# 14. FINAL GENERATION + SAVE
# ─────────────────────────────────────────────────────────────────────────────

def final_generation(cfg, tokenizer, history):
    model = JEPAGuidedTransformer(cfg).to(cfg.device)
    model.load_state_dict(torch.load("best_model_v5.pt", map_location=cfg.device))
    model.eval()

    print("\n── Final Generation ────────────────────────────────────────")
    results = []
    for prompt in cfg.gen_prompts:
        pid       = torch.tensor([tokenizer.encode(prompt).ids], dtype=torch.long)
        with_g    = decode(model.generate(pid, max_new=100, use_guidance=True),  tokenizer)
        without_g = decode(model.generate(pid, max_new=100, use_guidance=False), tokenizer)
        print(f"\n  Prompt : \"{prompt}\"")
        print(f"  [WITH]   {with_g}")
        print(f"  [WITHOUT]{without_g}")
        results.append({"prompt": prompt,
                        "with_guidance": with_g,
                        "without_guidance": without_g})

    out = {
        "version": "v5",
        "dataset": "wikitext-103",
        "model":   f"d_model={cfg.d_model}, n_layers={cfg.n_layers}, "
                   f"n_heads={cfg.n_heads}",
        "n_params": sum(p.numel() for p in
                        JEPAGuidedTransformer(cfg).parameters()
                        if p.requires_grad),
        "best_val_ppl_guided":  min(history["val_ppl_guided"]),
        "final_val_ppl_guided": history["val_ppl_guided"][-1],
        "final_val_ppl_base":   history["val_ppl_base"][-1],
        "final_guidance_delta": history["val_ppl_base"][-1] - history["val_ppl_guided"][-1],
        "generation_samples": results,
        "training_history": {
            "steps":          history["step"],
            "val_ppl_guided": history["val_ppl_guided"],
            "val_ppl_base":   history["val_ppl_base"],
            "val_align":      history["val_align"],
            "val_vc":         history["val_vc"],
            "lr":             history["lr"],
        },
        "config": {k: v for k, v in cfg.__dict__.items() if not callable(v)},
    }
    with open("jepa_ar_v5_results.json", "w") as f:
        json.dump(out, f, indent=2)
    print("\nResults → jepa_ar_v5_results.json")


# ─────────────────────────────────────────────────────────────────────────────
# 15. ENTRY POINT
# ─────────────────────────────────────────────────────────────────────────────

if __name__ == "__main__":
    t0 = time.time()

    # ── data ─────────────────────────────────────────────────────────────
    print("\n── Loading WikiText-103 ────────────────────────────────────")
    train_ds, val_ds = load_wikitext103()
    print(f"  Train rows: {len(train_ds):,}   Val rows: {len(val_ds):,}")

    print("\n── Building tokenizer ─────────────────────────────────────")
    tokenizer = build_tokenizer(train_ds, CFG.vocab_size)
    CFG.vocab_size = tokenizer.get_vocab_size()
    print(f"  Vocab size: {CFG.vocab_size}")

    print("\n── Tokenising & chunking ──────────────────────────────────")
    # Cap train rows on free T4: 200K rows ≈ 25M tokens ≈ 195K chunks (10x WikiText-2)
    # Set COLAB_PRO=True at top of file to use full corpus on A100
    MAX_TRAIN_ROWS = 200_000 if not COLAB_PRO else None
    train_chunks = tokenize_and_chunk(train_ds, tokenizer, CFG.seq_len,
                                      "train", max_rows=MAX_TRAIN_ROWS)
    val_chunks   = tokenize_and_chunk(val_ds,   tokenizer, CFG.seq_len, "val")
    print(f"  Train pairs: {len(train_chunks)-1:,}   Val pairs: {len(val_chunks)-1:,}")

    train_loader = DataLoader(PairDataset(train_chunks),
                              batch_size=CFG.batch_size, shuffle=True,
                              num_workers=2, pin_memory=True)
    val_loader   = DataLoader(PairDataset(val_chunks),
                              batch_size=CFG.batch_size, shuffle=False,
                              num_workers=2, pin_memory=True)

    # ── train ─────────────────────────────────────────────────────────────
    model, history = train(CFG, train_loader, val_loader, tokenizer)

    # ── results ───────────────────────────────────────────────────────────
    plot_training(history)
    final_generation(CFG, tokenizer, history)

    # ── ablation ──────────────────────────────────────────────────────────
    run_ablation(train_loader, val_loader, CFG, epochs=6)

    print(f"\nTotal time: {(time.time()-t0)/60:.1f} min")
    print("\nOutput files:")
    print("  best_model_v5.pt")
    print("  training_v5.png")
    print("  ablation_v5.png")
    print("  jepa_ar_v5_results.json")
    print(f"  {CFG.checkpoint_dir}/  (resume checkpoints)")
```

```python
"""
JEPA-Guided AR Transformer  —  v5_fast
========================================
Same architecture as v5 (validated in v4, scaled in v5).
Speed optimisations over v5:

  1. Mixed precision (torch.cuda.amp)        → ~45% faster on T4 Tensor Cores
  2. torch.compile                            → ~15% faster (PyTorch 2.0+)
  3. Batch size 32→64 (T4 has headroom)       → ~25% fewer optimizer steps
  4. DataLoader: workers=4, persistent,       → ~10% faster data pipeline
     pin_memory, prefetch_factor=2
  5. eval_every 500→1000, gen_every 1000→2000 → ~15% wall-time saved
  6. max_batches 100→50 in eval              → halves eval overhead
  7. 10 epochs (was 15)                       → converged by ~step 20K anyway

Expected runtime on free T4: ~45-55 min  (was ~100 min in v5)
Expected runtime on A100:    ~20-25 min

Architecture (unchanged — gradient isolation validated in v4):
  backbone → h
  h.detach() → jepa_proj → jepa_pred          (JEPA loss only)
  jepa_pred.mean().detach() → GuidanceCrossAttn(h,·) → correction
  h + correction → ar_head → ar_logits        (AR loss end-to-end)

Install:
    pip install torch datasets tokenizers matplotlib tqdm

Run:
    python jepa_ar_v5_fast.py
"""

import os, math, time, json, random, warnings
from copy import deepcopy
from dataclasses import dataclass, field
from pathlib import Path

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.cuda.amp import autocast, GradScaler
from torch.utils.data import Dataset, DataLoader
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from tqdm import tqdm

warnings.filterwarnings("ignore", category=UserWarning)

# ── reproducibility ───────────────────────────────────────────────────────────
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

# ─────────────────────────────────────────────────────────────────────────────
# HARDWARE FLAG  —  set True if you have Colab Pro / A100
# ─────────────────────────────────────────────────────────────────────────────
COLAB_PRO = False   # ← change to True for A100

# ─────────────────────────────────────────────────────────────────────────────
# 1.  CONFIG
# ─────────────────────────────────────────────────────────────────────────────
@dataclass
class Config:
    # ── model ─────────────────────────────────────────────────────────────
    vocab_size          : int   = 16000
    seq_len             : int   = 256 if COLAB_PRO else 128
    d_model             : int   = 256
    n_heads             : int   = 8
    n_layers            : int   = 6
    d_ff                : int   = 1024
    dropout             : float = 0.1

    # ── JEPA ─────────────────────────────────────────────────────────────
    jepa_dim            : int   = 128
    jepa_ema_decay      : float = 0.996
    jepa_lambda         : float = 1.0

    # ── VICReg ───────────────────────────────────────────────────────────
    vicreg_mu           : float = 1.0
    vicreg_nu           : float = 0.04
    vicreg_eps          : float = 1e-4

    # ── guidance cross-attention ──────────────────────────────────────────
    guidance_n_heads    : int   = 4
    guidance_init_scale : float = 0.0

    # ── inference repetition fix ──────────────────────────────────────────
    guidance_alpha_max  : float = 0.4
    guidance_alpha_min  : float = 0.1
    guidance_anneal_steps : int = 20

    # ── training ─────────────────────────────────────────────────────────
    # FAST: batch 64 on T4 (was 32) — T4 has 16GB and model is 9M params
    batch_size          : int   = 128 if COLAB_PRO else 64
    lr                  : float = 2e-4
    epochs              : int   = 10          # was 15 — converged by ~step 20K
    grad_clip           : float = 1.0
    warmup_steps        : int   = 1000
    lr_min              : float = 2e-5

    # ── speed flags ───────────────────────────────────────────────────────
    use_amp             : bool  = True        # mixed precision (fp16 forward)
    use_compile         : bool  = True        # torch.compile (PyTorch 2.0+)

    # ── checkpointing ─────────────────────────────────────────────────────
    save_every          : int   = 1000
    checkpoint_dir      : str   = "checkpoints_v5fast"
    resume              : bool  = True

    # ── eval / logging  (FAST: less frequent, fewer batches) ─────────────
    eval_every          : int   = 1000        # was 500
    gen_every           : int   = 2000        # was 1000
    eval_max_batches    : int   = 50          # was 100
    gen_prompts         : list  = field(default_factory=lambda: [
        "The history of science",
        "In the early 20th century",
    ])

    device : str = "cuda" if torch.cuda.is_available() else "cpu"


CFG = Config()
print(f"Device      : {CFG.device}")
print(f"d_model     : {CFG.d_model}  n_layers: {CFG.n_layers}  "
      f"seq_len: {CFG.seq_len}  batch: {CFG.batch_size}")
print(f"Mixed prec  : {CFG.use_amp}   compile: {CFG.use_compile}")
print(f"Dataset     : WikiText-103")
print(f"Epochs      : {CFG.epochs}")


# ─────────────────────────────────────────────────────────────────────────────
# 2.  DATA
# ─────────────────────────────────────────────────────────────────────────────

def load_wikitext103():
    from datasets import load_dataset
    print("  Downloading WikiText-103 (~500MB, first run only)...")
    ds = load_dataset("wikitext", "wikitext-103-raw-v1")
    return ds["train"], ds["validation"]


def build_tokenizer(train_ds, vocab_size=16000):
    from tokenizers import Tokenizer
    from tokenizers.models import BPE
    from tokenizers.trainers import BpeTrainer
    from tokenizers.pre_tokenizers import Whitespace
    import tempfile
    print(f"  Training BPE tokenizer on first 50K rows (vocab={vocab_size})...")
    tok = Tokenizer(BPE(unk_token="[UNK]"))
    tok.pre_tokenizer = Whitespace()
    trainer = BpeTrainer(vocab_size=vocab_size,
                         special_tokens=["[PAD]","[UNK]","[BOS]","[EOS]"])
    sample_rows = min(50_000, len(train_ds))
    with tempfile.NamedTemporaryFile("w", suffix=".txt", delete=False,
                                     encoding="utf-8") as f:
        for row in train_ds.select(range(sample_rows)):
            line = row["text"].strip()
            if line:
                f.write(line + "\n")
        tmp = f.name
    tok.train([tmp], trainer)
    os.unlink(tmp)
    return tok


def tokenize_and_chunk(ds_split, tokenizer, seq_len, desc="", max_rows=None):
    print(f"  Tokenising {desc} (row-by-row)...")
    all_ids = []
    rows = ds_split if max_rows is None else ds_split.select(range(max_rows))
    for row in tqdm(rows, desc=f"  tok {desc}", leave=False):
        text = row["text"].strip()
        if not text:
            continue
        all_ids.extend(tokenizer.encode(text).ids)
    print(f"    {desc}: {len(all_ids):,} tokens")
    n = (len(all_ids) // seq_len) * seq_len
    return np.array(all_ids[:n], dtype=np.int32).reshape(-1, seq_len)


class PairDataset(Dataset):
    def __init__(self, chunks):
        self.chunks = torch.tensor(chunks, dtype=torch.long)

    def __len__(self):
        return len(self.chunks) - 1

    def __getitem__(self, i):
        return self.chunks[i], self.chunks[i + 1]


# ─────────────────────────────────────────────────────────────────────────────
# 3.  MODEL COMPONENTS
# ─────────────────────────────────────────────────────────────────────────────

class RotaryEmbedding(nn.Module):
    def __init__(self, dim):
        super().__init__()
        inv_freq = 1.0 / (10000 ** (torch.arange(0, dim, 2).float() / dim))
        self.register_buffer("inv_freq", inv_freq)

    def forward(self, seq_len, device):
        t     = torch.arange(seq_len, device=device).float()
        freqs = torch.outer(t, self.inv_freq)
        emb   = torch.cat([freqs, freqs], dim=-1)
        return emb.cos(), emb.sin()


def rotate_half(x):
    x1, x2 = x.chunk(2, dim=-1)
    return torch.cat([-x2, x1], dim=-1)


def apply_rope(q, k, cos, sin):
    return q * cos + rotate_half(q) * sin, k * cos + rotate_half(k) * sin


class MultiHeadSelfAttention(nn.Module):
    def __init__(self, d_model, n_heads, dropout, causal=True):
        super().__init__()
        self.n_heads = n_heads
        self.d_head  = d_model // n_heads
        self.causal  = causal
        self.qkv     = nn.Linear(d_model, 3 * d_model, bias=False)
        self.out     = nn.Linear(d_model, d_model, bias=False)
        self.drop    = nn.Dropout(dropout)
        self.rope    = RotaryEmbedding(self.d_head)

    def forward(self, x):
        B, T, C = x.shape
        cos, sin = self.rope(T, x.device)
        cos = cos.unsqueeze(0).unsqueeze(0)
        sin = sin.unsqueeze(0).unsqueeze(0)
        qkv = self.qkv(x).reshape(B, T, 3, self.n_heads, self.d_head)
        q, k, v = qkv.unbind(2)
        q = q.transpose(1, 2); k = k.transpose(1, 2); v = v.transpose(1, 2)
        q, k = apply_rope(q, k, cos, sin)
        # Use F.scaled_dot_product_attention if available (PyTorch 2.0+)
        # — fused kernel, ~20% faster than manual matmul
        if hasattr(F, "scaled_dot_product_attention"):
            out = F.scaled_dot_product_attention(
                q, k, v,
                attn_mask=None,
                dropout_p=self.drop.p if self.training else 0.0,
                is_causal=self.causal,
            )
        else:
            scores = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(self.d_head)
            if self.causal:
                mask = torch.triu(torch.ones(T, T, device=x.device), diagonal=1).bool()
                scores = scores.masked_fill(mask, float("-inf"))
            out = self.drop(F.softmax(scores, dim=-1))
            out = torch.matmul(out, v)
        out = out.transpose(1, 2).reshape(B, T, C)
        return self.out(out)


class FFN(nn.Module):
    def __init__(self, d_model, d_ff, dropout):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_model, d_ff), nn.GELU(), nn.Dropout(dropout),
            nn.Linear(d_ff, d_model), nn.Dropout(dropout))

    def forward(self, x):
        return self.net(x)


class TransformerBlock(nn.Module):
    def __init__(self, d_model, n_heads, d_ff, dropout):
        super().__init__()
        self.attn = MultiHeadSelfAttention(d_model, n_heads, dropout)
        self.ffn  = FFN(d_model, d_ff, dropout)
        self.ln1  = nn.LayerNorm(d_model)
        self.ln2  = nn.LayerNorm(d_model)

    def forward(self, x):
        x = x + self.attn(self.ln1(x))
        x = x + self.ffn(self.ln2(x))
        return x


# ─────────────────────────────────────────────────────────────────────────────
# 4.  GUIDANCE CROSS-ATTENTION
# ─────────────────────────────────────────────────────────────────────────────

class GuidanceCrossAttention(nn.Module):
    """
    Cross-attention injecting JEPA future context into backbone representations.
    Q = backbone h  (B, T, d_model)
    K,V = jepa_summary  (B, 1, jepa_dim) → projected to d_model
    Output = residual correction, zero-initialised → starts as identity
    """
    def __init__(self, d_model, jepa_dim, n_heads, dropout, init_scale=0.0):
        super().__init__()
        assert d_model % n_heads == 0
        self.n_heads    = n_heads
        self.d_head     = d_model // n_heads
        self.scale      = math.sqrt(self.d_head)
        self.jepa_to_kv = nn.Linear(jepa_dim, 2 * d_model, bias=False)
        self.q_proj     = nn.Linear(d_model,  d_model,     bias=False)
        self.out_proj   = nn.Linear(d_model,  d_model,     bias=False)
        nn.init.constant_(self.out_proj.weight, init_scale)
        self.ln   = nn.LayerNorm(d_model)
        self.drop = nn.Dropout(dropout)

    def forward(self, h, jepa_summary, alpha=1.0):
        if alpha == 0.0:
            return torch.zeros_like(h)
        B, T, C = h.shape
        q  = self.q_proj(self.ln(h))
        q  = q.reshape(B, T, self.n_heads, self.d_head).transpose(1, 2)
        kv = self.jepa_to_kv(jepa_summary).unsqueeze(1)
        k, v = kv.chunk(2, dim=-1)
        k  = k.reshape(B, 1, self.n_heads, self.d_head).transpose(1, 2)
        v  = v.reshape(B, 1, self.n_heads, self.d_head).transpose(1, 2)
        scores = torch.matmul(q, k.transpose(-2, -1)) / self.scale
        attn   = self.drop(F.softmax(scores, dim=-1))
        out    = torch.matmul(attn, v).transpose(1, 2).reshape(B, T, C)
        return alpha * self.out_proj(out)


# ─────────────────────────────────────────────────────────────────────────────
# 5.  VICReg
# ─────────────────────────────────────────────────────────────────────────────

def vicreg_loss(z, cfg):
    if z.dim() == 3:
        B, T, D = z.shape; z = z.reshape(B * T, D)
    z = z - z.mean(dim=0)
    N, D = z.shape
    std      = torch.sqrt(z.var(dim=0) + cfg.vicreg_eps)
    var_loss = torch.relu(1.0 - std).mean()
    cov      = (z.T @ z) / (N - 1)
    off_diag = cov * (1 - torch.eye(D, device=z.device))
    cov_loss = off_diag.pow(2).sum() / D
    return cfg.vicreg_mu * var_loss + cfg.vicreg_nu * cov_loss


# ─────────────────────────────────────────────────────────────────────────────
# 6.  FULL MODEL
# ─────────────────────────────────────────────────────────────────────────────

class JEPAGuidedTransformer(nn.Module):
    def __init__(self, cfg: Config):
        super().__init__()
        self.cfg = cfg

        # backbone
        self.embed    = nn.Embedding(cfg.vocab_size, cfg.d_model)
        self.drop     = nn.Dropout(cfg.dropout)
        self.blocks   = nn.ModuleList([
            TransformerBlock(cfg.d_model, cfg.n_heads, cfg.d_ff, cfg.dropout)
            for _ in range(cfg.n_layers)
        ])
        self.ln_final = nn.LayerNorm(cfg.d_model)

        # AR head (weight-tied)
        self.ar_head        = nn.Linear(cfg.d_model, cfg.vocab_size, bias=False)
        self.ar_head.weight = self.embed.weight

        # JEPA online projector (backbone always detached before this)
        self.jepa_proj = nn.Sequential(
            nn.Linear(cfg.d_model, cfg.d_model),
            nn.GELU(),
            nn.Linear(cfg.d_model, cfg.jepa_dim),
        )

        # EMA target encoder
        self.target_embed  = deepcopy(self.embed)
        self.target_blocks = nn.ModuleList([deepcopy(b) for b in self.blocks])
        self.target_ln     = deepcopy(self.ln_final)
        self.target_proj   = deepcopy(self.jepa_proj)
        for p in self._target_params():
            p.requires_grad_(False)

        # Guidance cross-attention (between backbone and ar_head)
        self.guidance_xattn = GuidanceCrossAttention(
            d_model    = cfg.d_model,
            jepa_dim   = cfg.jepa_dim,
            n_heads    = cfg.guidance_n_heads,
            dropout    = cfg.dropout,
            init_scale = cfg.guidance_init_scale,
        )

        self._init_weights()

    def _target_params(self):
        for m in [self.target_embed, self.target_ln,
                  self.target_proj, *self.target_blocks]:
            yield from m.parameters()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, std=0.02)
                if m.bias is not None: nn.init.zeros_(m.bias)
            elif isinstance(m, nn.Embedding):
                nn.init.normal_(m.weight, std=0.02)
        nn.init.constant_(self.guidance_xattn.out_proj.weight,
                          self.cfg.guidance_init_scale)

    @torch.no_grad()
    def update_ema(self):
        d = self.cfg.jepa_ema_decay
        for online, target in [
            (self.embed,     self.target_embed),
            (self.ln_final,  self.target_ln),
            (self.jepa_proj, self.target_proj),
        ]:
            for po, pt in zip(online.parameters(), target.parameters()):
                pt.data.mul_(d).add_(po.data, alpha=1 - d)
        for bo, bt in zip(self.blocks, self.target_blocks):
            for po, pt in zip(bo.parameters(), bt.parameters()):
                pt.data.mul_(d).add_(po.data, alpha=1 - d)

    def _encode_backbone(self, x):
        h = self.drop(self.embed(x))
        for b in self.blocks: h = b(h)
        return self.ln_final(h)

    @torch.no_grad()
    def _encode_target(self, x):
        h = self.target_embed(x)
        for b in self.target_blocks: h = b(h)
        return self.target_ln(h)

    def forward(self, ctx, future=None, use_guidance=True, guidance_alpha=None):
        h         = self._encode_backbone(ctx)
        jepa_pred = self.jepa_proj(h.detach())      # backbone fully detached

        jepa_target = None
        if future is not None:
            with torch.no_grad():
                h_fut       = self._encode_target(future)
                jepa_target = self.target_proj(h_fut)

        if use_guidance:
            alpha        = guidance_alpha if guidance_alpha is not None \
                           else self.cfg.guidance_alpha_max
            jepa_summary = jepa_pred.mean(dim=1).detach()
            correction   = self.guidance_xattn(h, jepa_summary, alpha=alpha)
            h_guided     = h + correction
        else:
            h_guided = h

        ar_logits = self.ar_head(h_guided)
        return ar_logits, jepa_pred, jepa_target

    @torch.no_grad()
    def generate(self, prompt_ids, max_new=100, temperature=0.85,
                 top_k=50, top_p=0.92, use_guidance=True):
        self.eval()
        device = next(self.parameters()).device
        ids    = prompt_ids.to(device)
        cfg    = self.cfg

        for step in range(max_new):
            ctx = ids[:, -cfg.seq_len:]
            if use_guidance:
                t     = min(step / max(cfg.guidance_anneal_steps, 1), 1.0)
                alpha = cfg.guidance_alpha_max * (1-t) + cfg.guidance_alpha_min * t
            else:
                alpha = 0.0

            ar_logits, _, _ = self.forward(ctx, future=None,
                                           use_guidance=use_guidance,
                                           guidance_alpha=alpha)
            logits = ar_logits[:, -1, :] / temperature

            if ids.shape[1] > 1:
                for token_id in ids[0, -50:].unique():
                    logits[0, token_id] /= 1.3

            if top_k > 0:
                topk_v, _ = torch.topk(logits, min(top_k, logits.size(-1)))
                logits[logits < topk_v[:, -1:]] = float("-inf")

            if top_p < 1.0:
                sorted_logits, sorted_idx = torch.sort(logits, descending=True)
                cum_probs = torch.cumsum(F.softmax(sorted_logits, dim=-1), dim=-1)
                remove    = cum_probs - F.softmax(sorted_logits, dim=-1) > top_p
                sorted_logits[remove] = float("-inf")
                logits = torch.zeros_like(logits).scatter_(1, sorted_idx, sorted_logits)

            ids = torch.cat([ids, torch.multinomial(F.softmax(logits, -1), 1)], dim=1)

        self.train()
        return ids[0].tolist()


# ─────────────────────────────────────────────────────────────────────────────
# 7.  LOSS
# ─────────────────────────────────────────────────────────────────────────────

def compute_loss(ar_logits, jepa_pred, jepa_target, ctx_tokens, cfg):
    B, T, V = ar_logits.shape
    ar_loss = F.cross_entropy(
        ar_logits[:, :-1, :].reshape(-1, V),
        ctx_tokens[:, 1:].reshape(-1),
        ignore_index=0,
    )
    jepa_loss = torch.tensor(0.0, device=ar_logits.device)
    align_val = vc_val = 0.0
    if jepa_target is not None:
        pred_vec   = F.normalize(jepa_pred[:, -1, :],     dim=-1)
        target_vec = F.normalize(jepa_target.mean(dim=1), dim=-1)
        align_loss = F.smooth_l1_loss(pred_vec, target_vec)
        vc_loss    = vicreg_loss(jepa_pred, cfg)
        jepa_loss  = align_loss + vc_loss
        align_val  = align_loss.item()
        vc_val     = vc_loss.item()
    total = ar_loss + cfg.jepa_lambda * jepa_loss
    return total, ar_loss.item(), align_val, vc_val


# ─────────────────────────────────────────────────────────────────────────────
# 8.  LR SCHEDULE  —  linear warmup + cosine decay
# ─────────────────────────────────────────────────────────────────────────────

def get_lr(step, total_steps, cfg):
    if step < cfg.warmup_steps:
        return cfg.lr * step / max(1, cfg.warmup_steps)
    progress = (step - cfg.warmup_steps) / max(1, total_steps - cfg.warmup_steps)
    cosine   = 0.5 * (1 + math.cos(math.pi * progress))
    return cfg.lr_min + (cfg.lr - cfg.lr_min) * cosine


# ─────────────────────────────────────────────────────────────────────────────
# 9.  CHECKPOINT UTILS
# ─────────────────────────────────────────────────────────────────────────────

def save_checkpoint(model, optimizer, scaler, step, history, cfg):
    Path(cfg.checkpoint_dir).mkdir(exist_ok=True)
    path = os.path.join(cfg.checkpoint_dir, f"ckpt_step{step}.pt")
    # save underlying model if compiled
    m_state = (model._orig_mod if hasattr(model, "_orig_mod")
               else model).state_dict()
    torch.save({
        "step":      step,
        "model":     m_state,
        "optimizer": optimizer.state_dict(),
        "scaler":    scaler.state_dict() if scaler else None,
        "history":   history,
    }, path)
    ckpts = sorted(Path(cfg.checkpoint_dir).glob("ckpt_step*.pt"),
                   key=lambda p: int(p.stem.split("step")[1]))
    for old in ckpts[:-3]:
        old.unlink()
    return path


def load_latest_checkpoint(model, optimizer, scaler, cfg):
    ckpts = sorted(Path(cfg.checkpoint_dir).glob("ckpt_step*.pt"),
                   key=lambda p: int(p.stem.split("step")[1]))
    if not ckpts:
        return 0, None
    path = ckpts[-1]
    print(f"  Resuming from {path}")
    ckpt = torch.load(path, map_location=cfg.device)
    # load into underlying model if compiled
    m = model._orig_mod if hasattr(model, "_orig_mod") else model
    m.load_state_dict(ckpt["model"])
    optimizer.load_state_dict(ckpt["optimizer"])
    if scaler and ckpt.get("scaler"):
        scaler.load_state_dict(ckpt["scaler"])
    return ckpt["step"], ckpt["history"]


# ─────────────────────────────────────────────────────────────────────────────
# 10. EVAL
# ─────────────────────────────────────────────────────────────────────────────

@torch.no_grad()
def evaluate_both(model, loader, cfg):
    model.eval()
    ar_g = ar_b = n = 0
    V = cfg.vocab_size
    for i, (ctx, future) in enumerate(loader):
        if i >= cfg.eval_max_batches: break
        ctx, future = ctx.to(cfg.device, non_blocking=True), future.to(cfg.device, non_blocking=True)
        with autocast(enabled=cfg.use_amp):
            lg, _, _ = model(ctx, future, use_guidance=True)
            lg = lg.clone()          # ← add this
            lb, _, _ = model(ctx, future, use_guidance=False)
            lb = lb.clone()          # ← add this
        ar_g += F.cross_entropy(lg[:,:-1].reshape(-1,V),
                                ctx[:,1:].reshape(-1), ignore_index=0).item()
        ar_b += F.cross_entropy(lb[:,:-1].reshape(-1,V),
                                ctx[:,1:].reshape(-1), ignore_index=0).item()
        n += 1
    model.train()
    return math.exp(ar_g/n), math.exp(ar_b/n)


def decode(ids, tokenizer):
    return tokenizer.decode(ids)


def print_generation(model, tokenizer, cfg, step):
    # generation runs in fp32 for stability
    raw = model._orig_mod if hasattr(model, "_orig_mod") else model
    print(f"\n  ── Generation at step {step} ──")
    for prompt in cfg.gen_prompts:
        pid = torch.tensor([tokenizer.encode(prompt).ids], dtype=torch.long)
        with_g    = decode(raw.generate(pid, max_new=80, use_guidance=True),  tokenizer)
        without_g = decode(raw.generate(pid, max_new=80, use_guidance=False), tokenizer)
        print(f"\n  Prompt : \"{prompt}\"")
        print(f"  [WITH]   {with_g}")
        print(f"  [WITHOUT]{without_g}")


# ─────────────────────────────────────────────────────────────────────────────
# 11. TRAINING LOOP
# ─────────────────────────────────────────────────────────────────────────────

def train(cfg, train_loader, val_loader, tokenizer):
    print("\n── Building model ─────────────────────────────────────────")
    model = JEPAGuidedTransformer(cfg).to(cfg.device)
    n_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"  Trainable parameters: {n_params:,}")

    # ── torch.compile ────────────────────────────────────────────────────
    if cfg.use_compile and cfg.device == "cuda":
        print("  Compiling model with torch.compile (one-time, ~60s) ...")
        try:
            model = torch.compile(model, mode="reduce-overhead")
            print("  torch.compile: OK")
        except Exception as e:
            print(f"  torch.compile skipped: {e}")

    # ── mixed precision scaler ───────────────────────────────────────────
    scaler = GradScaler(enabled=(cfg.use_amp and cfg.device == "cuda"))

    optimizer = torch.optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=cfg.lr, weight_decay=0.01, betas=(0.9, 0.95))

    # ── resume ───────────────────────────────────────────────────────────
    start_step = 0
    history    = dict(step=[], train_ar=[], train_align=[], train_vc=[],
                      val_ppl_guided=[], val_ppl_base=[],
                      train_ppl=[], val_ppl=[], val_align=[], val_vc=[], lr=[])

    if cfg.resume:
        start_step, saved_history = load_latest_checkpoint(
            model, optimizer, scaler, cfg)
        if saved_history:
            history = saved_history
            print(f"  Resumed at step {start_step}, "
                  f"best val PPL so far: {min(history['val_ppl_guided']):.1f}")

    steps_per_epoch = len(train_loader)
    total_steps     = steps_per_epoch * cfg.epochs
    print(f"  Steps per epoch : {steps_per_epoch:,}")
    print(f"  Total steps     : {total_steps:,}")

    step         = start_step
    best_val_ppl = min(history["val_ppl_guided"]) if history["val_ppl_guided"] \
                   else float("inf")
    ep_ar = ep_align = ep_vc = ep_n = 0

    print("\n── Training ────────────────────────────────────────────────")

    for epoch in range(1, cfg.epochs + 1):
        ep_ar = ep_align = ep_vc = ep_n = 0
        pbar  = tqdm(train_loader, desc=f"Epoch {epoch}/{cfg.epochs}")

        for ctx, future in pbar:
            if step < start_step:
                step += 1; continue

            ctx    = ctx.to(cfg.device, non_blocking=True)
            future = future.to(cfg.device, non_blocking=True)
            lr     = get_lr(step, total_steps, cfg)
            for pg in optimizer.param_groups: pg["lr"] = lr

            # ── forward + backward with AMP ───────────────────────────────
            optimizer.zero_grad(set_to_none=True)   # faster than zero_grad()

            with autocast(enabled=cfg.use_amp):
                ar_logits, jp, jt = model(ctx, future, use_guidance=True)
                loss, ar_l, al_l, vc_l = compute_loss(ar_logits, jp, jt, ctx, cfg)

            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.grad_clip)
            scaler.step(optimizer)
            scaler.update()

            # EMA update on underlying model weights
            raw = model._orig_mod if hasattr(model, "_orig_mod") else model
            raw.update_ema()

            ep_ar += ar_l; ep_align += al_l; ep_vc += vc_l
            ep_n  += 1; step += 1

            pbar.set_postfix({
                "ar":    f"{ar_l:.3f}",
                "align": f"{al_l:.4f}",
                "ppl":   f"{math.exp(min(ar_l, 10)):.1f}",
                "lr":    f"{lr:.2e}",
            })

            # ── eval ──────────────────────────────────────────────────────
            if step % cfg.eval_every == 0:
                ppl_guided, ppl_base = evaluate_both(model, val_loader, cfg)
                train_ppl = math.exp(min(ep_ar / ep_n, 10))
                delta = ppl_base - ppl_guided

                history["step"].append(step)
                history["train_ar"].append(ep_ar / ep_n)
                history["train_align"].append(ep_align / ep_n)
                history["train_vc"].append(ep_vc / ep_n)
                history["val_ppl_guided"].append(ppl_guided)
                history["val_ppl_base"].append(ppl_base)
                history["train_ppl"].append(train_ppl)
                history["val_ppl"].append(ppl_guided)
                history["val_align"].append(ep_align / ep_n)
                history["val_vc"].append(ep_vc / ep_n)
                history["lr"].append(lr)

                if ppl_guided < best_val_ppl:
                    best_val_ppl = ppl_guided
                    raw = model._orig_mod if hasattr(model, "_orig_mod") else model
                    torch.save(raw.state_dict(), "best_model_v5fast.pt")

                print(f"\n  Step {step:6d} | "
                      f"Train PPL {train_ppl:.1f} | "
                      f"Val guided {ppl_guided:.1f} | "
                      f"base {ppl_base:.1f} | "
                      f"Δ {delta:+.2f} | "
                      f"lr {lr:.2e}")

            # ── generation ────────────────────────────────────────────────
            if step % cfg.gen_every == 0:
                print_generation(model, tokenizer, cfg, step)

            # ── checkpoint ────────────────────────────────────────────────
            if step % cfg.save_every == 0:
                path = save_checkpoint(model, optimizer, scaler,
                                       step, history, cfg)
                print(f"  Checkpoint → {path}")

    raw = model._orig_mod if hasattr(model, "_orig_mod") else model
    torch.save(raw.state_dict(), "best_model_v5fast.pt")
    return model, history


# ─────────────────────────────────────────────────────────────────────────────
# 12. ABLATION
# ─────────────────────────────────────────────────────────────────────────────

def run_ablation(train_loader, val_loader, cfg, epochs=4):
    """
    Lighter ablation than v5 — 4 epochs, 20% data subset.
    Uses AMP for speed. Checks: AR-only, JEPA no guidance, JEPA guided.
    """
    print(f"\n── Ablation: AR-only vs JEPA-guided ({epochs} epochs) ─────")
    MAX_BATCHES = len(train_loader) // 5

    def train_variant(use_jepa, use_guidance, label):
        cfg2 = deepcopy(cfg)
        cfg2.use_compile = False           # don't compile short runs
        if not use_jepa:
            cfg2.jepa_lambda = 0.0

        m   = JEPAGuidedTransformer(cfg2).to(cfg.device)
        opt = torch.optim.AdamW(
            [p for p in m.parameters() if p.requires_grad],
            lr=cfg2.lr, weight_decay=0.01, betas=(0.9, 0.95))
        sc  = GradScaler(enabled=(cfg2.use_amp and cfg.device == "cuda"))
        steps_tot = MAX_BATCHES * epochs
        ppls = []
        s    = 0
        for ep in range(epochs):
            count = 0
            for ctx, future in tqdm(train_loader, desc=f"  {label} ep{ep+1}",
                                    leave=False, total=MAX_BATCHES):
                if count >= MAX_BATCHES: break
                ctx    = ctx.to(cfg.device, non_blocking=True)
                future = future.to(cfg.device, non_blocking=True)
                lr     = get_lr(s, steps_tot, cfg2)
                for pg in opt.param_groups: pg["lr"] = lr
                opt.zero_grad(set_to_none=True)
                with autocast(enabled=cfg2.use_amp):
                    ar_logits, jp, jt = m(ctx, future, use_guidance=use_guidance)
                    loss, *_ = compute_loss(ar_logits, jp, jt, ctx, cfg2)
                sc.scale(loss).backward()
                sc.unscale_(opt)
                torch.nn.utils.clip_grad_norm_(m.parameters(), cfg2.grad_clip)
                sc.step(opt); sc.update()
                m.update_ema()
                count += 1; s += 1

            # eval after each epoch
            ppl_g, ppl_b = evaluate_both(m, val_loader, cfg2)
            ppls.append(ppl_g if use_guidance else ppl_b)
            print(f"    {label} ep{ep+1}: PPL {ppls[-1]:.2f}")
        return ppls

    ar_ppls     = train_variant(False, False, "AR-only       ")
    jepa_base   = train_variant(True,  False, "JEPA no-guide ")
    jepa_guided = train_variant(True,  True,  "JEPA guided   ")

    delta_guidance = jepa_base[-1]  - jepa_guided[-1]
    delta_vs_ar    = ar_ppls[-1]    - jepa_guided[-1]
    print(f"\n  ── Ablation Results ──")
    print(f"  AR-only              : {ar_ppls[-1]:.2f}")
    print(f"  JEPA no guidance     : {jepa_base[-1]:.2f}")
    print(f"  JEPA guided          : {jepa_guided[-1]:.2f}")
    print(f"  Guidance benefit     : {delta_guidance:+.2f}  "
          f"({'✓' if delta_guidance > 0 else '✗'})")
    print(f"  vs AR-only           : {delta_vs_ar:+.2f}  "
          f"({'✓ JEPA beats AR' if delta_vs_ar > 0 else '✗ AR still better'})")

    ep_range = list(range(1, epochs + 1))
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    fig.suptitle("v5_fast Ablation — WikiText-103 subset", fontsize=12)
    axes[0].plot(ep_range, ar_ppls,     "o-",  label="AR only",          color="#E05C5C", lw=2)
    axes[0].plot(ep_range, jepa_guided, "s-",  label="JEPA guided",      color="#4A90D9", lw=2)
    axes[0].plot(ep_range, jepa_base,   "^--", label="JEPA no guidance", color="#85C1E9", lw=1.5)
    axes[0].set_title("Val PPL ↓"); axes[0].set_xlabel("Epoch")
    axes[0].legend(fontsize=8); axes[0].grid(True, alpha=0.3)
    deltas_g  = [b - g for b, g in zip(jepa_base, jepa_guided)]
    deltas_ar = [ar - jg for ar, jg in zip(ar_ppls, jepa_guided)]
    axes[1].bar(ep_range, deltas_g, color=["#4A90D9" if d>0 else "#E05C5C" for d in deltas_g])
    axes[1].axhline(0, color="black", lw=0.8)
    axes[1].set_title("Guidance benefit (base−guided) ↑"); axes[1].grid(True, alpha=0.3, axis="y")
    axes[2].bar(ep_range, deltas_ar, color=["#4A90D9" if d>0 else "#E05C5C" for d in deltas_ar])
    axes[2].axhline(0, color="black", lw=0.8)
    axes[2].set_title("JEPA guided vs AR-only ↑"); axes[2].grid(True, alpha=0.3, axis="y")
    plt.tight_layout()
    plt.savefig("ablation_v5fast.png", dpi=150)
    print("Ablation plot → ablation_v5fast.png")
    return ar_ppls, jepa_guided, jepa_base


# ─────────────────────────────────────────────────────────────────────────────
# 13. PLOTS
# ─────────────────────────────────────────────────────────────────────────────

def plot_training(history):
    fig, axes = plt.subplots(2, 3, figsize=(16, 8))
    fig.suptitle("JEPA-Guided AR Transformer v5_fast — WikiText-103", fontsize=13)
    s = history["step"]

    def plot(ax, tk, vk, title, ct="#4A90D9", cv="#E05C5C"):
        ax.plot(s, history[tk], label="Train", color=ct, lw=1.5)
        ax.plot(s, history[vk], label="Val",   color=cv, lw=1.5)
        ax.set_title(title); ax.set_xlabel("Step")
        ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

    plot(axes[0,0], "train_ppl",   "val_ppl",   "Perplexity (guided) ↓")
    plot(axes[0,1], "train_ar",    "train_ar",  "AR Loss ↓")
    plot(axes[0,2], "train_align", "val_align", "JEPA Align Loss ↓",
         ct="#8E44AD", cv="#6C3483")
    plot(axes[1,0], "train_vc",    "val_vc",    "VICReg Loss ↓",
         ct="#E67E22", cv="#CA6F1E")

    ax = axes[1, 1]
    ax.plot(s, history["val_ppl_guided"], label="Val PPL (guided)", color="#4A90D9", lw=2)
    ax.plot(s, history["val_ppl_base"],   label="Val PPL (base)",   color="#85C1E9",
            lw=1.5, linestyle="--")
    deltas = [b - g for b, g in zip(history["val_ppl_base"], history["val_ppl_guided"])]
    ax2 = ax.twinx()
    ax2.plot(s, deltas, color="#27AE60", lw=1, linestyle=":", alpha=0.7, label="Δ (base−guided)")
    ax2.set_ylabel("Δ PPL", fontsize=8, color="#27AE60")
    ax.set_title("Guidance Effect on Val PPL")
    ax.set_xlabel("Step"); ax.legend(fontsize=7, loc="upper right"); ax.grid(True, alpha=0.3)

    ax = axes[1, 2]
    if history["lr"]:
        ax.plot(s, history["lr"], color="#27AE60", lw=1.5)
        ax.set_title("Learning Rate (cosine decay)")
        ax.set_xlabel("Step"); ax.grid(True, alpha=0.3)
        ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x:.1e}"))

    plt.tight_layout()
    plt.savefig("training_v5fast.png", dpi=150)
    print("Training plot → training_v5fast.png")


# ─────────────────────────────────────────────────────────────────────────────
# 14. FINAL GENERATION + SAVE
# ─────────────────────────────────────────────────────────────────────────────

def final_generation(cfg, tokenizer, history):
    model = JEPAGuidedTransformer(cfg).to(cfg.device)
    model.load_state_dict(
        torch.load("best_model_v5fast.pt", map_location=cfg.device))
    model.eval()

    print("\n── Final Generation ────────────────────────────────────────")
    results = []
    for prompt in [
        "The history of science",
        "In the early 20th century",
        "The political system of",
        "During the Second World War",
    ]:
        pid       = torch.tensor([tokenizer.encode(prompt).ids], dtype=torch.long)
        with_g    = decode(model.generate(pid, max_new=120, use_guidance=True),  tokenizer)
        without_g = decode(model.generate(pid, max_new=120, use_guidance=False), tokenizer)
        print(f"\n  Prompt : \"{prompt}\"")
        print(f"  [WITH]   {with_g}")
        print(f"  [WITHOUT]{without_g}")
        results.append({"prompt": prompt,
                        "with_guidance": with_g,
                        "without_guidance": without_g})

    out = {
        "version": "v5_fast",
        "dataset": "wikitext-103",
        "model":   f"d_model={cfg.d_model}, n_layers={cfg.n_layers}, n_heads={cfg.n_heads}",
        "n_params": sum(p.numel() for p in JEPAGuidedTransformer(cfg).parameters()
                        if p.requires_grad),
        "best_val_ppl_guided":  min(history["val_ppl_guided"]) if history["val_ppl_guided"] else None,
        "final_val_ppl_guided": history["val_ppl_guided"][-1]  if history["val_ppl_guided"] else None,
        "final_val_ppl_base":   history["val_ppl_base"][-1]    if history["val_ppl_base"]   else None,
        "final_guidance_delta": (history["val_ppl_base"][-1] - history["val_ppl_guided"][-1])
                                 if history["val_ppl_guided"] else None,
        "generation_samples": results,
        "training_history": {
            "steps":          history["step"],
            "val_ppl_guided": history["val_ppl_guided"],
            "val_ppl_base":   history["val_ppl_base"],
            "lr":             history["lr"],
        },
        "config": {k: v for k, v in cfg.__dict__.items() if not callable(v)},
    }
    with open("jepa_ar_v5fast_results.json", "w") as f:
        json.dump(out, f, indent=2)
    print("\nResults → jepa_ar_v5fast_results.json")


# ─────────────────────────────────────────────────────────────────────────────
# 15. ENTRY POINT
# ─────────────────────────────────────────────────────────────────────────────

if __name__ == "__main__":
    t0 = time.time()

    print("\n── Loading WikiText-103 ────────────────────────────────────")
    train_ds, val_ds = load_wikitext103()
    print(f"  Train rows: {len(train_ds):,}   Val rows: {len(val_ds):,}")

    print("\n── Building tokenizer ─────────────────────────────────────")
    tokenizer = build_tokenizer(train_ds, CFG.vocab_size)
    CFG.vocab_size = tokenizer.get_vocab_size()
    print(f"  Vocab size: {CFG.vocab_size}")

    print("\n── Tokenising & chunking ──────────────────────────────────")
    MAX_TRAIN_ROWS = 200_000 if not COLAB_PRO else None
    train_chunks = tokenize_and_chunk(train_ds, tokenizer, CFG.seq_len,
                                      "train", max_rows=MAX_TRAIN_ROWS)
    val_chunks   = tokenize_and_chunk(val_ds,   tokenizer, CFG.seq_len, "val")
    print(f"  Train pairs: {len(train_chunks)-1:,}   Val pairs: {len(val_chunks)-1:,}")

    # FAST: workers=4, persistent_workers, prefetch_factor=2
    train_loader = DataLoader(
        PairDataset(train_chunks),
        batch_size=CFG.batch_size, shuffle=True,
        num_workers=4, pin_memory=True,
        persistent_workers=True, prefetch_factor=2,
    )
    val_loader = DataLoader(
        PairDataset(val_chunks),
        batch_size=CFG.batch_size, shuffle=False,
        num_workers=4, pin_memory=True,
        persistent_workers=True, prefetch_factor=2,
    )

    model, history = train(CFG, train_loader, val_loader, tokenizer)

    plot_training(history)
    final_generation(CFG, tokenizer, history)
    run_ablation(train_loader, val_loader, CFG, epochs=4)

    print(f"\nTotal time: {(time.time()-t0)/60:.1f} min")
    print("\nOutput files:")
    print("  best_model_v5fast.pt")
    print("  training_v5fast.png")
    print("  ablation_v5fast.png")
    print("  jepa_ar_v5fast_results.json")
    print(f"  {CFG.checkpoint_dir}/  (resume checkpoints)")
```
